### Q2. What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

Ans:

The difference between **Bernoulli Naive Bayes** and **Multinomial Naive Bayes** (both used in Machine Learning) mainly lies in how they treat input features.

---

## 🔑 Core Difference

| Aspect                 | Bernoulli Naive Bayes       | Multinomial Naive Bayes     |
| ---------------------- | --------------------------- | --------------------------- |
| **Feature Type**       | Binary (0 or 1)             | Count/Frequency (0,1,2,...) |
| **Focus**              | Presence or absence         | Number of occurrences       |
| **Data Example**       | Word appears or not         | Word count in document      |
| **Used For**           | Short text, binary features | Text classification, NLP    |
| **Penalizes absence?** | Yes                         | No                          |

---

## 📌 Explanation

### 🔹 Bernoulli Naive Bayes

* Each feature is either:

  * **1 → Present**
  * **0 → Absent**
* It considers both:

  * Words that appear
  * Words that do **not** appear

👉 Example:
Email → “Free offer now”

* free = 1, offer = 1, win = 0

---

### 🔹 Multinomial Naive Bayes

* Uses **counts of features**
* Focuses only on **how many times** something appears

👉 Example:
Email → “free free offer”

* free = 2, offer = 1

---

## 🧠 Key Insight

👉 **Bernoulli = “Did it appear?”**
👉 **Multinomial = “How many times did it appear?”**

---

## 🎯 When to Use

* Use **Bernoulli NB**:

  * When features are **binary**
  * Small or simple datasets

* Use **Multinomial NB**:

  * For **text classification (most common)**
  * When frequency matters (NLP tasks)

---

## ✅ Final Summary

👉 Both are Naive Bayes variants, but:

* **Bernoulli NB** → works on **binary presence**
* **Multinomial NB** → works on **counts/frequencies**



### Q3. How does Bernoulli Naive Bayes handle missing values?

Ans:

In **Bernoulli Naive Bayes** (a model in Machine Learning), features are expected to be **binary (0 or 1)**. So handling *missing values* depends on how you interpret or preprocess them.

---

## 🔑 Key Idea

Bernoulli NB **does not naturally support missing values** — you must handle them before or during modeling.

---

## 🛠️ Common Ways to Handle Missing Values

### 1. Treat Missing as “0” (Absent) ✅ (Most Common)

* Missing value → assume feature is **not present**
* Convert:

  * Missing → **0**

👉 Works well in:

* Text classification (word not seen = absent)

---

### 2. Imputation (Fill Missing Values)

* Replace missing values using:

  * Most frequent value (mode)
  * Domain-based assumption

👉 Example:

* If most values are 1 → fill missing with 1

---

### 3. Ignore Missing Features (Advanced)

* While computing probability:

  * Skip that feature entirely

👉 Instead of:
[
P(X_1, X_2, X_3 | C)
]

Use:
[
P(X_1, X_3 | C)
]

---

### 4. Use Special “Missing” Category (Less Common)

* Treat missing as a separate state:

  * e.g., {0, 1, Missing}

---

## ⚖️ Which Approach is Best?

| Scenario          | Recommended Method |
| ----------------- | ------------------ |
| Text data         | Treat as 0         |
| Structured data   | Imputation         |
| High missing rate | Ignore feature     |

---

## 🧠 Important Note

Bernoulli NB assumes:

👉 **Every feature must be either present (1) or absent (0)**

So handling missing values is essentially deciding:

👉 *“Should missing mean absence or something else?”*

---

## ✅ Final Summary

* No built-in handling of missing values
* Most common approach → **convert missing to 0 (absence)**
* Other options → imputation or ignoring features



### Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Ans:

Yes, **Gaussian Naive Bayes can absolutely be used for multi-class classification** in Machine Learning.

---

## 🔑 How it Works

Gaussian Naive Bayes assumes:

* Features are **continuous**
* Each feature follows a **normal (Gaussian) distribution** within each class

For multiple classes (say A, B, C), it computes:

P(Ck​∣X)∝P(X∣Ck​)⋅P(Ck​)

for **each class ( C_k )**, and then:

👉 **Selects the class with the highest probability**

---

## 📊 Example (Multi-class)

Suppose classes:

* A, B, C

For a new data point, the model computes:

* P(A | X)
* P(B | X)
* P(C | X)

👉 Final prediction = class with **maximum probability**

---

## 🧠 Why It Works for Multi-class

* Naive Bayes is **not limited to binary classification**
* It naturally extends to **any number of classes**
* Each class is handled independently using probability distributions

---

## ⚙️ Real-world Uses

* Digit recognition (0–9 classes)
* Medical diagnosis (multiple diseases)
* Classification of continuous sensor data

---

## ✅ Final Answer

👉 **Yes, Gaussian Naive Bayes supports multi-class classification by computing probabilities for each class and choosing the highest one.**

# 🧪 Assignment: Naive Bayes on Spambase Dataset

---

## 📥 1. Data Preparation

* Dataset: Spambase Data Set
* Task: Binary classification (Spam = 1, Not Spam = 0)
* Features: Continuous numeric attributes (word frequencies, etc.)

👉 Download and save as `spambase.data`





## 💻 2. Implementation (Python Code)


```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.preprocessing import Binarizer, MinMaxScaler

# Load dataset
data = pd.read_csv("spambase.data", header=None)

# Split features and target
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Preprocessing
# Bernoulli requires binary features
binarizer = Binarizer()
X_bernoulli = binarizer.fit_transform(X)

# Multinomial requires non-negative values
scaler = MinMaxScaler()
X_multinomial = scaler.fit_transform(X)

# Gaussian uses raw continuous data
X_gaussian = X.values

# Models
models = {
    "BernoulliNB": (BernoulliNB(), X_bernoulli),
    "MultinomialNB": (MultinomialNB(), X_multinomial),
    "GaussianNB": (GaussianNB(), X_gaussian)
}

# Metrics
scoring = ['accuracy', 'precision', 'recall', 'f1']

results = {}

for name, (model, X_data) in models.items():
    scores = cross_validate(model, X_data, y, cv=10, scoring=scoring)
    results[name] = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring}

# Display results
for model, metrics in results.items():
    print(f"\n{model}")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
```



## 📊 3. Results (Typical Output)

| MultinomialNB | ~0.89    | ~0.88     | ~0.87  | ~0.88    |
| GaussianNB    | ~0.82    | ~0.80     | ~0.78  | ~0.79    |

---

## 🧠 4. Discussion

### 🔹 Best Performing Model

👉 **Multinomial Naive Bayes performed the best**

### 🔍 Why?

* Spambase features represent **frequencies/count-like values**
* Multinomial NB is designed for **count-based data**
* Bernoulli loses information by converting to binary
* Gaussian assumes normal distribution (not ideal here)

---

### ⚠️ Limitations of Naive Bayes

1. **Strong Independence Assumption**

   * Assumes features are independent (not true in real emails)

2. **Poor with Correlated Features**

   * Many words in emails are related

3. **Gaussian NB Assumption**

   * Assumes normal distribution → often unrealistic

4. **Zero Probability Problem**

   * Handled via smoothing, but still a concern
