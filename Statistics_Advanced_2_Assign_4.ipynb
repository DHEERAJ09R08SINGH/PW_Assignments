### Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio test. The function should return the F-value and the corresponding p-value for the test.

```python
import numpy as np
from scipy.stats import f

def variance_ratio_test(sample1, sample2):
    """
    Performs an F-test to compare variances of two samples.

    Parameters:
    sample1, sample2 : array-like
        Input data samples

    Returns:
    F_value : float
        Calculated F statistic
    p_value : float
        Two-tailed p-value
    """

    # Convert to numpy arrays
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)

    # Sample variances (ddof=1 for sample variance)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Ensure larger variance is numerator
    if var1 >= var2:
        F_value = var1 / var2
    else:
        F_value = var2 / var1
        df1, df2 = df2, df1

    # Two-tailed p-value
    p_value = 2 * (1 - f.cdf(F_value, df1, df2))

    return F_value, p_value
```

```python
data1 = [12, 15, 14, 10, 13, 17, 16]
data2 = [22, 25, 24, 20, 23, 27, 26]

F_stat, p_val = variance_ratio_test(data1, data2)

print("F-value:", F_stat)
print("p-value:", p_val)
```

### Interpretation

* If p-value < α (e.g., 0.05) → Reject the null hypothesis
→ Variances are significantly different.

* If p-value ≥ α → Fail to reject the null hypothesis
→ No significant difference in variances.

### Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an F-distribution, write a Python function that returns the critical F-value for a two-tailed test.

Ans:

For a two-tailed F-test at significance level α=0.05, we split the significance level into two parts:
* α/2=0.025 in each tail

Since the F-distribution is right-skewed and non-symmetric, the two critical values are:
* Upper critical value:
F((1−α)/2)
	​
* Lower critical value:
F(α/2)

  ​
We use the percent point function (ppf) from scipy.stats.f.


```python
from scipy.stats import f

def critical_f_value(alpha, df1, df2):
    """
    Returns lower and upper critical F-values for a two-tailed test.

    Parameters:
    alpha : float
        Significance level (e.g., 0.05)
    df1 : int
        Degrees of freedom for numerator
    df2 : int
        Degrees of freedom for denominator

    Returns:
    (F_lower, F_upper) : tuple
        Lower and upper critical F-values
    """

    # Split alpha for two-tailed test
    alpha_half = alpha / 2

    # Lower critical value
    F_lower = f.ppf(alpha_half, df1, df2)

    # Upper critical value
    F_upper = f.ppf(1 - alpha_half, df1, df2)

    return F_lower, F_upper
```

```python
alpha = 0.05
df1 = 10   # numerator degrees of freedom
df2 = 15   # denominator degrees of freedom

F_lower, F_upper = critical_f_value(alpha, df1, df2)

print("Lower Critical F-value:", F_lower)
print("Upper Critical F-value:", F_upper)
```

### Interpretation Rule

For a two-tailed F-test:

* Reject H0 if

   * F< F_lower or F>F_upper
	​


Otherwise, fail to reject H0.


### Q3. Write a Python program that generates random samples from two normal distributions with known variances and uses an F-test to determine if the variances are equal. The program should output the F-value, degrees of freedom, and p-value for the test.

```python
import numpy as np
from scipy.stats import f

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for normal distributions
n1 = 30
n2 = 35
mean1, variance1 = 50, 25   # variance = 25
mean2, variance2 = 55, 36   # variance = 36

# Generate random samples
sample1 = np.random.normal(mean1, np.sqrt(variance1), n1)
sample2 = np.random.normal(mean2, np.sqrt(variance2), n2)

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Degrees of freedom
df1 = n1 - 1
df2 = n2 - 1

# Ensure larger variance is numerator
if var1 >= var2:
    F_value = var1 / var2
else:
    F_value = var2 / var1
    df1, df2 = df2, df1

# Two-tailed p-value
p_value = 2 * (1 - f.cdf(F_value, df1, df2))

# Output results
print("F-value:", F_value)
print("Degrees of Freedom (df1, df2):", (df1, df2))
print("p-value:", p_value)
```

### Interpretation

* If p-value < 0.05 → Reject H0
   * Variances are significantly different.

* If p-value ≥ 0.05 → Fail to reject H0
   * No significant difference in variances.

### Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from each population. Conduct an F-test at the 5% significance level to determine if the variances are significantly different.

Ans:

We are given:

* Population variance 1: (α_1^2 = 10 )
* Population variance 2: (α_2^2 = 15 )
* Sample sizes: ( n_1 = n_2 = 12 )
* Significance level: ( α = 0.05 )

---

### Step 1: State Hypotheses


* H_0: α_1^2 = α_2^2

* H_a: α_1^2 != α_2^2

This is a **two-tailed F-test**.


### Step 2: Compute Test Statistic

Since 15 is larger, place it in the numerator:

F = 15/10 = 1.5


### Step 3: Degrees of Freedom

* df_1 = n_1 - 1 = 11

* df_2 = n_2 - 1 = 11


So,

df_1 = 11, df_2 = 11

### Step 4: Critical F-Values (Two-Tailed, α = 0.05)

For a two-tailed test:

* α/2 = 0.025

From F-tables (or using Python):

* F_(upper) ≈ 3.59

Lower critical value:

* F_(lower) = 1/F_(upper) ≈ 0.28

### Step 5: Decision Rule

Reject ( H_0 ) if:
* F < 0.28 or F > 3.59

Our test statistic:
* F = 1.5

Since:
0.28 < 1.5 < 3.59


---

### Conclusion

At the 5% significance level:

* We **fail to reject** the null hypothesis.

There is **no significant difference** between the two population variances.

This is a **two-tailed test**.

### Step 2: Test Statistic

For testing a single variance, we use:

F = (s^2)/(α_0^2)

F = 0.006 / 0.005 = 1.2

### Step 3: Degrees of Freedom

* df_1 = n - 1 = 24

For a single-variance F-test, we compare:

* F(24, infty)

* α/2 = 0.005

From F tables:
* F_(upper)≈ 2.80

* F = 1.2

---

### Final Conclusion

At the **1% significance level**, we **fail to reject** the null hypothesis.

### Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an F-distribution and calculates the mean and variance of the distribution. The function should return the mean and variance as a tuple.

Ans:

For an F-distribution with:

* Numerator degrees of freedom → d1
	​
* Denominator degrees of freedom → d2
	​
The formulas are:

* Mean
  * Mean
    = (d2 / 𝑑2−2) ,for 𝑑2>2
	​
* Variance

     * Variance=2𝑑2^2(𝑑1+𝑑2−2)/𝑑1(𝑑2−2)^2(𝑑2−4),for 𝑑2>4

⚠ Conditions:

Mean exists only if
* d2>2

Variance exists only if
* d2>4

```python
def f_distribution_stats(df1, df2):
    """
    Calculates mean and variance of an F-distribution.

    Parameters:
    df1 : int
        Degrees of freedom for numerator
    df2 : int
        Degrees of freedom for denominator

    Returns:
    (mean, variance) : tuple
        Mean and variance of the F-distribution
        (returns None if not defined)
    """

    # Mean
    if df2 > 2:
        mean = df2 / (df2 - 2)
    else:
        mean = None  # Mean does not exist

    # Variance
    if df2 > 4:
        variance = (2 * (df2 ** 2) * (df1 + df2 - 2)) / \
                   (df1 * ((df2 - 2) ** 2) * (df2 - 4))
    else:
        variance = None  # Variance does not exist

    return mean, variance
```

```python
df1 = 10
df2 = 15

mean, variance = f_distribution_stats(df1, df2)

print("Mean:", mean)
print("Variance:", variance)
```
