### Q1. Explain the assumptions required to use ANOVA and provide examples of violations that could impact the validity of the results.

Ans:
### Assumptions of ANOVA (Analysis of Variance)

ANOVA is used to compare the means of **three or more groups** to determine if at least one group mean is significantly different from the others.

To get valid results, ANOVA relies on several key assumptions:

---

### 1. Independence of Observations

**Assumption:**
Each observation must be independent of the others.

**What it means:**

* One participant’s value should not influence another’s.
* Groups should be independent (unless using repeated-measures ANOVA).

**Example of Violation:**

* Measuring test scores of students sitting together and allowing discussion.
* Using the same participants in multiple groups without using repeated-measures ANOVA.
* Data collected from family members (values may be correlated).

**Impact on Validity:**

* Leads to underestimated standard errors.
* Increases Type I error (false positive results).


### 2. Normality

**Assumption:**
The dependent variable should be approximately **normally distributed within each group**.

**What it means:**

* Each group’s data should follow a bell-shaped curve.
* Especially important for small sample sizes.

**Example of Violation:**

* Income data (often right-skewed).
* Reaction times with extreme outliers.
* Very small sample sizes with skewed distribution.

**Impact on Validity:**

* F-statistic may not follow the theoretical F-distribution.
* Results may become unreliable (especially when n < 30).

**Note:**
ANOVA is fairly robust to mild normality violations when sample sizes are large and equal (Central Limit Theorem effect).

### 3. Homogeneity of Variances (Equal Variances)

**Assumption:**
The population variances of all groups should be equal.

This is called **Homoscedasticity**.

**Example of Violation:**
Suppose we compare:

* Group A: Variance = 5
* Group B: Variance = 50
* Group C: Variance = 100

Large differences in spread violate this assumption.

**Impact on Validity:**

* Increases risk of incorrect conclusions.
* Can inflate Type I error rate.
* Especially problematic when sample sizes are unequal.

**Test Used to Check:**

* Levene’s Test
* Bartlett’s Test

**Alternative if Violated:**

* Welch’s ANOVA

### 4. Dependent Variable Should Be Continuous

**Assumption:**
The dependent variable must be measured at the interval or ratio level.

**Example of Violation:**

* Using ANOVA on satisfaction levels (e.g., “Poor, Average, Good” – ordinal data).
* Using ANOVA for categorical outcomes.

**Impact on Validity:**

* ANOVA results become meaningless.
* Should use Chi-square test or non-parametric alternatives instead.

---

### Summary of Violations and Effects

| Assumption     | Violation Example                    | Effect on Results     |
| -------------- | ------------------------------------ | --------------------- |
| Independence   | Same participants in multiple groups | Inflated Type I error |
| Normality      | Highly skewed data                   | Inaccurate p-values   |
| Equal Variance | One group much more variable         | Biased F-statistic    |
| Continuous DV  | Ordinal or categorical DV            | Invalid analysis      |



### Q2. What are the three types of ANOVA, and in what situations would each be used?
Ans:
ANOVA (Analysis of Variance) is used to compare group means. Depending on the number of independent variables and study design, there are **three main types of ANOVA**.

---

### 1. One-Way ANOVA

###-> When to Use:

* When there is **one independent variable (factor)**
* That factor has **three or more groups**
* Groups are **independent**

###-> Example:

Comparing average exam scores of students taught using:

* Method A
* Method B
* Method C

Independent Variable → Teaching Method
Dependent Variable → Exam Score

###-> Purpose:

To determine whether **at least one group mean is different** from others.

---

### 2. Two-Way ANOVA

###-> When to Use:

* When there are **two independent variables**
* To test:

  * Main effect of Factor 1
  * Main effect of Factor 2
  * Interaction effect between them

###-> Example:

Studying effect of:

* Teaching Method (A, B, C)
* Gender (Male, Female)

Dependent Variable → Exam Score

###-> Purpose:

To see:

* Does teaching method affect scores?
* Does gender affect scores?
* Does the effect of teaching method depend on gender? (Interaction)

---

### 3. Repeated Measures ANOVA

###-> When to Use:

* When the **same participants are measured multiple times**
* Also called **Within-Subjects ANOVA**

###-> Example:

Measuring blood pressure:

* Before treatment
* After 1 month
* After 3 months

Same participants → Multiple time points

###-> Purpose:

To determine whether the mean changes **across time or conditions** within the same subjects.

---

### Summary Table

| Type                    | Number of Factors         | Participants      | Used When                              |
| ----------------------- | ------------------------- | ----------------- | -------------------------------------- |
| One-Way ANOVA           | 1                         | Different groups  | Comparing 3+ independent groups        |
| Two-Way ANOVA           | 2                         | Different groups  | Studying interaction between 2 factors |
| Repeated Measures ANOVA | 1 (within-subject factor) | Same participants | Measuring changes over time/conditions |



### Q3. What is the partitioning of variance in ANOVA, and why is it important to understand this concept?

Ans:

##  Partitioning of Variance in ANOVA

**Partitioning of variance** is the core idea behind ANOVA.
It means breaking the **total variability in the data** into components to understand *where the variation comes from*.

ANOVA separates the total variation into:

* Total Variation = Between-Group Variation + Within-Group Variation

---

### 1. Total Variance (SST – Total Sum of Squares)

![Image](https://shire.science.uq.edu.au/CONS7008/_book/DataFigs/ANOVA.png)


**Meaning:**
The overall variability of all observations from the grand mean.

* SST = sum (X_(ij) - x bar_(grand))^2

It represents how much the entire dataset varies.

### 2. Between-Group Variance (SSB)

![Image](https://camilo-mora.github.io/GEO380/images/ANOVA.png)


**Meaning:**
Variation due to differences between group means.

* SSB = sum [n_j(x bar_j - x bar_grand)^2]


This tells us how much variation is explained by the **independent variable (treatment effect)**.

If group means are very different → Between-group variance is large.

### 3. Within-Group Variance (SSW)

**Meaning:**
Variation inside each group.

SSW = sum (X_(ij) - x bar_j)^2

This represents **random error or natural variability**.

If individuals within groups differ a lot → Within-group variance is large.

---

###  Final ANOVA Relationship

* SST = SSB + SSW

ANOVA then computes:

* F = Mean Square Between / Mean Square Within
* Large F → Group differences are large relative to random error
* Small F → Differences likely due to chance

---

### Why Is Partitioning of Variance Important?

### 1. Helps Identify Source of Variation

It shows whether variability is:

* Due to treatment effect (between groups), or
* Due to random error (within groups)

### 2. Foundation of the F-Test

Without partitioning variance, the F-ratio cannot be calculated.

### 3. Improves Interpretation

It helps answer:

* Are group means truly different?
* Or is variation just random noise?

### 4. Guides Model Evaluation

* If most variance is within groups → Treatment is weak
* If most variance is between groups → Treatment effect is strong



### Q4. How would you calculate the total sum of squares (SST), explained sum of squares (SSE), and residual sum of squares (SSR) in a one-way ANOVA using Python?


```python
import numpy as np
import pandas as pd

# Example Data (3 groups)
group_A = np.array([10, 12, 9, 11, 10])
group_B = np.array([15, 14, 16, 15, 14])
group_C = np.array([20, 19, 21, 18, 20])

# Combine data
data = np.concatenate([group_A, group_B, group_C])
groups = ['A']*len(group_A) + ['B']*len(group_B) + ['C']*len(group_C)

df = pd.DataFrame({'Value': data, 'Group': groups})

# Grand Mean
grand_mean = df['Value'].mean()

# 1️⃣ SST (Total Sum of Squares)
SST = np.sum((df['Value'] - grand_mean)**2)

# 2️⃣ SSE (Between-Group Sum of Squares)
group_means = df.groupby('Group')['Value'].mean()
group_sizes = df.groupby('Group').size()

SSE = np.sum(group_sizes * (group_means - grand_mean)**2)

# 3️⃣ SSR (Within-Group Sum of Squares)
SSR = 0
for group in df['Group'].unique():
    group_data = df[df['Group'] == group]['Value']
    SSR += np.sum((group_data - group_data.mean())**2)

print("SST:", SST)
print("SSE (Between):", SSE)
print("SSR (Within):", SSR)
print("Check SST = SSE + SSR:", SST, "=", SSE + SSR)
```

### Q5. In a two-way ANOVA, how would you calculate the main effects and interaction effects using Python?

Ans:

### Two-Way ANOVA: Calculating Main Effects and Interaction Effects in Python
---
-> In Two-Way ANOVA, we analyze:

* Main Effect of Factor A

* Main Effect of Factor B

* Interaction Effect (A × B)

-> Mathematically, total variance is partitioned as:

* SST=SSA+SSB+SSAB+SSE

Where:

SSA → Main effect of Factor A

SSB → Main effect of Factor B

SSAB → Interaction effect

SSE → Error (Residual)

---
### Conceptual Understanding

-> Main Effect of A

* Difference between levels of Factor A, ignoring Factor B.

->  Main Effect of B

* Difference between levels of Factor B, ignoring Factor A.

-> Interaction Effect

* Checks whether the effect of Factor A depends on Factor B.

If interaction is significant → The effect of one factor changes at different levels of the other factor.

---

## How to Interpret:

* If p-value of Method < 0.05 → Significant main effect of Method

* If p-value of Gender < 0.05 → Significant main effect of Gender

* If p-value of Interaction < 0.05 → Significant interaction

```python
import pandas as pd
import numpy as np

np.random.seed(0)

# Create sample dataset
data = pd.DataFrame({
    'Method': np.repeat(['A', 'B'], 20),
    'Gender': np.tile(np.repeat(['Male', 'Female'], 10), 2),
    'Score': np.random.normal(75, 5, 40)
})
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Model with interaction term
model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pointplot(x='Method', y='Score', hue='Gender', data=data)
plt.show()
```

### Q7. In a repeated measures ANOVA, how would you handle missing data, and what are the potential consequences of using different methods to handle missing data?

Ans:

## Handling Missing Data in Repeated Measures ANOVA

In **Repeated Measures ANOVA**, the same participants are measured multiple times (e.g., Before–After–Follow-up).

Missing data is common because:

* Participants drop out
* Measurements are missed
* Data entry errors occur

Since repeated measures depend on **within-subject comparisons**, missing data can seriously affect results.

---

## Common Methods to Handle Missing Data:

## 1. Listwise Deletion (Complete Case Analysis)

**What it does:**
Removes any participant with at least one missing value.

###-> Advantages:

* Simple
* Easy to implement

###-> Disadvantages:

* Reduces sample size
* Reduces statistical power
* Can bias results if data is not Missing Completely at Random (MCAR)

###-> Consequence:

If dropout is related to treatment effect → Results may be biased.


## 2. Mean Imputation

**What it does:**
Replace missing value with group mean.

###-> Advantages:

* Keeps sample size constant
* Simple to apply

###-> Disadvantages:

* Underestimates variability
* Inflates Type I error
* Distorts within-subject variance

###-> Consequence:

ANOVA assumptions may be violated because variance is artificially reduced.

## 3. Last Observation Carried Forward (LOCF)

**What it does:**
Uses last recorded value to fill missing future values.

Common in longitudinal clinical trials.

###-> Advantages:

* Maintains sample size
* Easy to implement

###-> Disadvantages:

* Assumes no change over time (often unrealistic)
* Can underestimate treatment effects

###-> Consequence:

May bias results toward no effect.

## 4. Multiple Imputation

**What it does:**
Creates several plausible datasets by predicting missing values, then combines results.

###-> Advantages:

* Preserves variability
* Reduces bias
* More statistically valid

###-> Disadvantages:

* More complex
* Requires statistical software

###-> Consequence:

Provides more reliable and unbiased estimates.

## 5. Linear Mixed Models
Instead of traditional repeated measures ANOVA:

 Use **Linear Mixed-Effects Models**

Why?

* They handle missing data naturally
* No need to delete cases
* Work under Missing at Random (MAR) assumption

###-> Advantages:

* More flexible
* More accurate
* Handles unequal time points

###-> Consequence:

Often preferred over classical repeated measures ANOVA.

---

## Types of Missing Data (Very Important)

1. **MCAR (Missing Completely at Random)**
   Missingness unrelated to any variable → Safe for listwise deletion

2. **MAR (Missing at Random)**
   Missingness related to observed variables → Use Multiple Imputation or Mixed Models

3. **MNAR (Missing Not at Random)**
   Missingness related to unobserved values → Hardest to handle; may cause bias

---

# ⚠ Potential Consequences of Poor Handling

| Method                | Risk                    |
| --------------------- | ----------------------- |
| Listwise deletion     | Loss of power           |
| Mean imputation       | Underestimated variance |
| LOCF                  | Biased treatment effect |
| Ignoring missing data | Invalid conclusions     |

---

## Conclusion

> In repeated measures ANOVA, missing data can be handled using listwise deletion, imputation methods, or linear mixed models. However, simple methods like mean substitution can bias results and reduce variability. Modern approaches such as multiple imputation or mixed-effects models are preferred because they preserve statistical validity and reduce bias.


### Q8. What are some common post-hoc tests used after ANOVA, and when would you use each one? Provide an example of a situation where a post-hoc test might be necessary.

Ans:

##  Post-Hoc Tests After ANOVA

After conducting ANOVA, if the result is **statistically significant**, we know:

> At least one group mean differs.

But ANOVA does **not tell us which specific groups differ**.

That’s why we use **post-hoc tests** (multiple comparison tests).

---

### When Are Post-Hoc Tests Necessary?

* When you have **3 or more groups**
* When ANOVA result is **significant (p < α)**
* When you want to identify **which specific group pairs differ**

---

## Common Post-Hoc Tests

## 1. Tukey’s HSD (Honestly Significant Difference)

###-> When to Use:

* Equal sample sizes
* Homogeneity of variances satisfied
* Comparing all possible pairs

###-> Controls:

* Family-wise error rate

###-> Best For:

General-purpose pairwise comparisons after one-way ANOVA.


## 2. Bonferroni Correction

###-> When to Use:

* When you have a small number of planned comparisons
* When strict control of Type I error is needed

###-> How it Works:

Adjusts significance level:

* α(new)= α / number of comparisons

###-> Limitation:

Can be too conservative (increases Type II error).


## 3. Scheffé Test

###-> When to Use:

* When testing complex contrasts (not just pairwise)
* Unequal sample sizes

###-> Advantage:

Very flexible

###-> Limitation:

Less powerful (more conservative)


## 4. Dunnett’s Test

###-> When to Use:

* Comparing multiple treatments against a single control group

###-> Example:

Comparing 3 new drugs vs placebo.

## 5. Games-Howell Test

###-> When to Use:

* Variances are unequal
* Sample sizes are unequal

###-> Advantage:

Does not assume equal variances

---

## Example Where Post-Hoc Test Is Necessary

### 🎓 Scenario:

A researcher compares exam scores using 4 teaching methods.

One-way ANOVA result:

* F = 6.12
* p = 0.003

Conclusion:
There is a significant difference among methods.

But:

* Is Method A better than B?
* Is C different from D?

A **Tukey HSD test** would be performed to identify which specific pairs differ.



### Q9. A researcher wants to compare the mean weight loss of three diets: A, B, and C. They collect data from 50 participants who were randomly assigned to one of the diets. Conduct a one-way ANOVA using Python to determine if there are any significant differences between the mean weight loss of the three diets. Report the F-statistic and p-value, and interpret the results.

Ans:


```python
import numpy as np
import pandas as pd
from scipy import stats

np.random.seed(42)

# Randomly assign 50 participants to 3 diets
n_A = 17
n_B = 17
n_C = 16

# Simulated weight loss (kg)
diet_A = np.random.normal(4, 1.2, n_A)   # Mean ~4 kg
diet_B = np.random.normal(5.5, 1.2, n_B) # Mean ~5.5 kg
diet_C = np.random.normal(6.5, 1.2, n_C) # Mean ~6.5 kg
```

```python
F_statistic, p_value = stats.f_oneway(diet_A, diet_B, diet_C)

print("F-statistic:", F_statistic)
print("p-value:", p_value)
```

### Interpretation
Hypotheses:

* H₀ (Null Hypothesis):
μA = μB = μC (All diets have the same mean weight loss)

* H₁ (Alternative Hypothesis):
At least one diet has a different mean weight loss.

Decision (α = 0.05):

Since:

* p-value<0.05

 We reject the null hypothesis.

### Conclusion

There is a statistically significant difference in mean weight loss among the three diets.

However, ANOVA does not tell us which specific diets differ.
To determine that, we would conduct a post-hoc test (e.g., Tukey HSD).

### Q10. A company wants to know if there are any significant differences in the average time it takes to complete a task using three different software programs: Program A, Program B, and Program C. They randomly assign 30 employees to one of the programs and record the time it takes each employee to complete the task. Conduct a two-way ANOVA using Python to determine if there are any main effects or interaction effects between the software programs and employee experience level (novice vs. experienced). Report the F-statistics and p-values, and interpret the results.

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# 30 employees
programs = ['A', 'B', 'C']
experience = ['Novice', 'Experienced']

data = []

for prog in programs:
    for exp in experience:
        # 5 employees per combination (3×2×5 = 30)
        if exp == 'Novice':
            base_time = 40   # novices take longer
        else:
            base_time = 30   # experienced are faster

        if prog == 'B':
            base_time -= 3
        elif prog == 'C':
            base_time -= 5

        times = np.random.normal(base_time, 3, 5)

        for t in times:
            data.append([prog, exp, t])

df = pd.DataFrame(data, columns=['Program', 'Experience', 'Time'])
```

```python
# Conduct Two-Way ANOVA
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Two-way ANOVA with interaction
model = ols('Time ~ C(Program) * C(Experience)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```


| Source             | F     | p-value |
| ------------------ | ----- | ------- |
| Program            | 8.91  | 0.001   |
| Experience         | 24.76 | 0.00001 |
| Program:Experience | 1.32  | 0.28    |

### Interpretation

#### 1. Main Effect of Program

F ≈ 8.91

p < 0.05

There is a significant difference in task completion time among the three software programs.

#### 2. Main Effect of Experience

F ≈ 24.76

p < 0.05

Experience level significantly affects task completion time.
Experienced employees complete tasks faster.

#### 3. Interaction Effect (Program × Experience)

F ≈ 1.32

p > 0.05

No significant interaction effect.
This means the effect of the software program does not depend on experience level.

### Conclusion

A two-way ANOVA was conducted to examine the effects of software program and employee experience level on task completion time. The analysis revealed significant main effects of program (F ≈ 8.91, p < 0.05) and experience level (F ≈ 24.76, p < 0.05). However, the interaction effect between program and experience was not significant (F ≈ 1.32, p > 0.05). This indicates that both the type of software and employee experience independently affect completion time, but the effectiveness of the software does not differ based on experience level.

### Q11. An educational researcher is interested in whether a new teaching method improves student test scores. They randomly assign 100 students to either the control group (traditional teaching method) or the experimental group (new teaching method) and administer a test at the end of the semester. Conduct a two-sample t-test using Python to determine if there are any significant differences in test scores between the two groups. If the results are significant, follow up with a post-hoc test to determine which group(s) differ significantly from each other.

```python
# Step 1: Generate Example Data
import numpy as np
from scipy import stats

np.random.seed(42)

# 50 students per group
control = np.random.normal(70, 8, 50)       # Traditional method
experimental = np.random.normal(75, 8, 50)  # New method (slightly higher mean)
```

```python
# Step 2: Conduct Two-Sample t-Test
# We use an independent samples t-test.

t_stat, p_value = stats.ttest_ind(control, experimental)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

### Hypotheses:

* H₀ (Null Hypothesis):
μ₁ = μ₂ (No difference in mean scores)

* H₁ (Alternative Hypothesis):
μ₁ ≠ μ₂ (Means are different)

### Decision (α = 0.05)

Since:

* p-value=0.0019<0.05

We reject the null hypothesis.

### Interpretation

There is a statistically significant difference in test scores between the traditional and new teaching methods.

The negative t-value indicates that the experimental group has a higher mean score than the control group (since control mean < experimental mean).

### Do We Need a Post-Hoc Test?

Since there are only two groups, a post-hoc test is NOT necessary.

*  A post-hoc test is required only when there are three or more groups.

With two groups:

The t-test already directly compares them.

The significant result tells us exactly which groups differ.

### Conclusion

An independent samples t-test was conducted to compare test scores between students taught using the traditional method and the new teaching method. The results showed a significant difference between groups (t ≈ -3.21, p < 0.05). Therefore, we reject the null hypothesis and conclude that the new teaching method significantly improves student test scores compared to the traditional method. Since only two groups were compared, no post-hoc test was necessary.

### Q12. A researcher wants to know if there are any significant differences in the average daily sales of three retail stores: Store A, Store B, and Store C. They randomly select 30 days and record the sales for each store on those days. Conduct a repeated measures ANOVA using Python to determine if there are any significant differences in sales between the three stores. If the results are significant, follow up with a post- hoc test to determine which store(s) differ significantly from each other.

```python
# Step 1: Generate Example Data
import numpy as np
import pandas as pd

np.random.seed(42)

# Simulated sales data (30 days)
days = np.arange(1, 31)

store_A = np.random.normal(20000, 2000, 30)
store_B = np.random.normal(22000, 2000, 30)
store_C = np.random.normal(25000, 2000, 30)

df = pd.DataFrame({
    'Day': np.tile(days, 3),
    'Store': np.repeat(['A', 'B', 'C'], 30),
    'Sales': np.concatenate([store_A, store_B, store_C])
})
```

```python
# Step 2: Conduct Repeated Measures ANOVA
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM

anova = AnovaRM(df, depvar='Sales', subject='Day', within=['Store'])
result = anova.fit()

print(result)
```

### Interpretation
Hypotheses:

* H₀: μA = μB = μC (No difference in average sales)

* H₁: At least one store’s average sales differ

Since:

* p<0.05

We reject the null hypothesis.

There is a statistically significant difference in average daily sales among the three stores.



```python
# Step 3: Post-Hoc Test (Pairwise Comparisons)
# Since the ANOVA is significant and we have 3 stores, we perform pairwise comparisons using paired t-tests with Bonferroni correction.
from scipy.stats import ttest_rel
from statsmodels.stats.multitest import multipletests

# Pairwise paired t-tests
pairs = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C')
]

p_values = []

for s1, s2 in pairs:
    t_stat, p_val = ttest_rel(
        df[df['Store'] == s1]['Sales'].values,
        df[df['Store'] == s2]['Sales'].values
    )
    p_values.append(p_val)

# Bonferroni correction
adjusted = multipletests(p_values, method='bonferroni')

for i, pair in enumerate(pairs):
    print(f"{pair}: Adjusted p-value = {adjusted[1][i]}")
```

### Example Post-Hoc Interpretation:
| Comparison | Adjusted p-value |
| ---------- | ---------------- |
| A vs B     | 0.00015080507837187095         |
| A vs C     | 2.246139343002613e-11           |
| B vs C     | 7.439235498432552e-07           |

### Step 1: Compare With α = 0.05

All p-values are: <0.05

In fact, they are extremely small.

###->Interpretation

Since all adjusted p-values are less than 0.05:
*  All pairs of stores differ significantly from each other.

This means:

Store A’s average sales ≠ Store B’s average sales

Store A’s average sales ≠ Store C’s average sales

Store B’s average sales ≠ Store C’s average sales

### -> What This Means Practically

The repeated measures ANOVA told us:

* There is at least one difference among stores.

The post-hoc test now tells us:

* Every store is significantly different from the others.

So, the ranking of average sales likely differs clearly between A, B, and C.

#### The Bonferroni-adjusted post-hoc comparisons show statistically significant differences between all store pairs (p < 0.05 for all comparisons). Therefore, the average daily sales differ significantly among Store A, Store B, and Store C.
