### Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans:

Estimation Statistics (a part of inferential statistics) is the process of using sample data to estimate or infer the value of an unknown population parameter.

In real life, we usually cannot study the entire population (due to time, cost, or practicality). So, we collect a sample and use it to estimate population values like:

* Population mean (Î¼)

* Population proportion (p)

* Population variance (ÏƒÂ²)

---

1. Point Estimate

A point estimate is a single value used to estimate an unknown population parameter.

It gives one specific number as the best guess.

-> Example:

If the average marks of 50 sampled students is 72, then:

xË‰=72

Here, 72 is the point estimate of the population mean (Î¼).

| Population Parameter      | Point Estimate (Sample Statistic) |
| ------------------------- | --------------------------------- |
| Î¼ (population mean)       | xÌ„ (sample mean)                  |
| p (population proportion) | pÌ‚ (sample proportion)            |
| ÏƒÂ² (population variance)  | sÂ² (sample variance)              |

-> Advantage:

Simple and easy to calculate.

-> Limitation:

Does not tell how accurate or reliable the estimate is.

---

2. Interval Estimate


An interval estimate gives a range of values within which the population parameter is likely to lie.

This range is called a Confidence Interval (CI).

ðŸ“Œ Example:

Instead of saying the average marks is 72, we say:

(68,76)

This means we are confident that the population mean lies between 68 and 76.

If it is a 95% confidence interval, it means:

We are 95% confident that the true population mean lies in this interval.

### General Formula for Confidence Interval (Mean):

Point Estimate Â± Margin of Error

xÌ„ Â± Z(s/sqrt(n))

Where:

xÌ„ = sample mean

Z = Z-value (depends on confidence level)

s = sample standard deviation

n = sample size



### Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.


```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    """
    Estimates the population mean using confidence interval.

    Parameters:
    sample_mean (float): Sample mean (xÌ„)
    sample_std (float): Sample standard deviation (s)
    sample_size (int): Sample size (n)
    confidence_level (float): Confidence level (default = 0.95)

    Returns:
    tuple: (lower_bound, upper_bound)
    """

    # Z-scores for common confidence levels
    z_values = {
        0.90: 1.645,
        0.95: 1.96,
        0.99: 2.576
    }

    if confidence_level not in z_values:
        raise ValueError("Supported confidence levels: 0.90, 0.95, 0.99")

    z = z_values[confidence_level]

    margin_of_error = z * (sample_std / math.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
```

```python
sample_mean = 72
sample_std = 10
sample_size = 50

ci = estimate_population_mean(sample_mean, sample_std, sample_size)

print("95% Confidence Interval:", ci)
```

### Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.


```python
import numpy as np
from scipy import stats

# Sample data from two populations
sample1 = np.array([72, 75, 78, 70, 69, 74, 71])
sample2 = np.array([65, 68, 70, 66, 64, 67, 69])

# Perform independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Significance level
alpha = 0.05

print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis (Significant difference exists)")
else:
    print("Fail to Reject the Null Hypothesis (No significant difference)")
```

### Q8. Define p-value and explain its significance in hypothesis testing.

The **p-value (probability value)** is the probability of obtaining a test statistic as extreme as (or more extreme than) the observed value, **assuming the null hypothesis (Hâ‚€) is true**.

### Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

We are given:
* Two independent random samples
* Equal sample sizes
* Null hypothesis:
   H0â€‹:Î¼1â€‹=Î¼2â€‹
We will use an Independent Two-Sample t-test.

```python
import numpy as np
from scipy import stats

# Generate two random samples of equal size
np.random.seed(42)

sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=52, scale=5, size=30)

# Null Hypothesis: H0: Î¼1 = Î¼2
# Perform independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Significance level
alpha = 0.05

print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis (Means are significantly different)")
else:
    print("Fail to Reject the Null Hypothesis (No significant difference)")
```

### Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

Ans:

### Given Data

* Sample size ( n = 50 )
* Sample mean ( \bar{x} = 500 ) dollars
* Sample standard deviation ( s = 50 ) dollars
* Confidence level = 95%

Since:

* Sample size is large (n â‰¥ 30)
* Population standard deviation is unknown

We use the **t-distribution** (though for n=50, Z and t give very similar results).

---

### Step 1: Formula for Confidence Interval

xÌ„ Â± t(s/sqrt(n))

For 95% confidence and df = 49,
( t â‰ˆ 2.009 )

### Step 2: Calculate Standard Error

SE = s/sqrt(n) = 50/7.071 =7.07


### Step 3: Calculate Margin of Error

ME=tÃ—SE
ME=2.009Ã—7.07
MEâ‰ˆ14.21

### Step 4: Construct Confidence Interval

500Â±14.21

Lower limit:
500 - 14.21 = 485.79

Upper limit:
500 + 14.21 = 514.21

---

### Final Answer

The **95% confidence interval** for the population mean revenue is:

(485.79,; 514.21)

---

### Interpretation

We are **95% confident** that the true average daily revenue of the coffee shop lies between **$485.79 and $514.21**.



```python
import math
from scipy import stats

# Given data
sample_mean = 500
sample_std = 50
n = 50
confidence_level = 0.95

# Degrees of freedom
df = n - 1

# Standard Error
standard_error = sample_std / math.sqrt(n)

# t critical value
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Margin of Error
margin_of_error = t_critical * standard_error

# Confidence Interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("95% Confidence Interval:")
print(f"({lower_bound:.2f}, {upper_bound:.2f})")
```

### Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

Ans:

Here is the Python code to test the hypothesis using a one-sample t-test.

#### Given:

* Hypothesized mean (Î¼â‚€) = 10 mmHg
* Sample mean (xÌ„) = 8 mmHg
* Sample standard deviation (s) = 3 mmHg
* Sample size (n) = 100
* Significance level (Î±) = 0.05

```python
import numpy as np
from scipy import stats
import math

# Given values
mu_0 = 10          # Hypothesized mean decrease
sample_mean = 8    # Observed sample mean
sample_std = 3     # Sample standard deviation
n = 100            # Sample size
alpha = 0.05       # Significance level

# Calculate t-statistic manually
standard_error = sample_std / math.sqrt(n)
t_statistic = (sample_mean - mu_0) / standard_error

# Degrees of freedom
df = n - 1

# Calculate p-value (two-tailed test)
p_value = 2 * stats.t.cdf(t_statistic, df=df)

print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 6))

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

### Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.

Ans:

 Here we perform a one-sample Z-test because the population standard deviation (Ïƒ = 0.5) is known.

#### Given:

* Population mean (Î¼â‚€) = 5 pounds
* Population standard deviation (Ïƒ) = 0.5 pounds
* Sample size (n) = 25
* Sample mean (xÌ„) = 4.8 pounds
* Significance level (Î±) = 0.01
* Left-tailed test

Hypotheses:

H0â€‹:Î¼=5

H1â€‹:Î¼<5


```python
import math
from scipy import stats

# Given values
mu_0 = 5          # Hypothesized population mean
sigma = 0.5       # Population standard deviation
sample_mean = 4.8 # Sample mean
n = 25            # Sample size
alpha = 0.01      # Significance level

# Calculate Standard Error
standard_error = sigma / math.sqrt(n)

# Calculate Z-statistic
z_statistic = (sample_mean - mu_0) / standard_error

# Calculate p-value (left-tailed test)
p_value = stats.norm.cdf(z_statistic)

print("Z-statistic:", round(z_statistic, 4))
print("P-value:", round(p_value, 6))

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

### Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.

Ans:

We will perform an independent two-sample t-test (since population standard deviations are unknown).

### Given:

* ð‘›1=30 , ð‘¥Ë‰1=80, ð‘ 1=10

* ð‘›2=40, ð‘¥Ë‰2=75,ð‘ 2=8

* Significance level  Î±=0.01

Hypotheses:
  * H0â€‹:Î¼1â€‹=Î¼2â€‹
  * H1â€‹:Î¼1â€‹î€ =Î¼2â€‹
  (Two-tailed test)


```python
import math
from scipy import stats

# Given values
n1 = 30
n2 = 40
mean1 = 80
mean2 = 75
s1 = 10
s2 = 8
alpha = 0.01

# Calculate pooled variance
pooled_variance = (((n1 - 1)*(s1**2)) + ((n2 - 1)*(s2**2))) / (n1 + n2 - 2)

# Standard error
standard_error = math.sqrt(pooled_variance * ((1/n1) + (1/n2)))

# t-statistic
t_statistic = (mean1 - mean2) / standard_error

# Degrees of freedom
df = n1 + n2 - 2

# Two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 6))

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```
