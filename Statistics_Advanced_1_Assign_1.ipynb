### Q2. What are the types of Probability distribution?

Ans:

## Types of Probability Distribution

Probability distributions are mainly classified into **two types**:


### 1. Discrete Probability Distribution

Used when the random variable takes **countable values** (0, 1, 2, 3, â€¦).

It uses a **Probability Mass Function (PMF)**.

#### Common Types of Discrete Distributions :

##### 1. **Binomial Distribution**


* Fixed number of trials
* Each trial has two outcomes (success/failure)
* Example: Tossing a coin 5 times

##### 2. **Poisson Distribution**

* Used for counting number of events in fixed time/space
* Example: Number of calls in 1 hour

##### 3. **Geometric Distribution**

* Number of trials until first success

### 2. Continuous Probability Distribution

Used when the random variable takes **infinite values within an interval**.

It uses a **Probability Density Function (PDF)**.

#### Common Types of Continuous Distributions

##### 1. **Normal Distribution**

* Bell-shaped and symmetric
* Defined by mean (Î¼) and standard deviation (Ïƒ)

##### 2. **Uniform Distribution**

* All values in interval are equally likely

##### 3. **Exponential Distribution**

* Models waiting time between events


### Summary Table

| Type       | Used For         | Function Type |
| ---------- | ---------------- | ------------- |
| Discrete   | Countable values | PMF           |
| Continuous | Infinite values  | PDF           |


### Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.


```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF)
    of a normal distribution at a given point.

    Parameters:
    x : float -> point at which PDF is calculated
    mean : float -> mean (Î¼)
    std_dev : float -> standard deviation (Ïƒ)

    Returns:
    float -> PDF value at x
    """

    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

    return coefficient * exponent
```

```python
result = normal_pdf(x=2, mean=0, std_dev=1)
print(result)
```

### Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.
Ans:
### Binomial distribution

The **Binomial Distribution** is a discrete probability distribution that models the number of successes in a fixed number of independent trials.

#### Properties of Binomial Distribution

1. **Fixed number of trials (n)**

* The experiment is repeated a fixed number of times.

2.  **Only two possible outcomes**

* Each trial results in:

  * Success (with probability **p**)
  * Failure (with probability **q = 1 âˆ’ p**)

3.  **Constant probability**

* Probability of success (**p**) remains the same for every trial.

4.  **Independent trials**

* Outcome of one trial does not affect another.

5.  **Random Variable (X)**

* X represents the number of successes in n trials.
* ( X = 0, 1, 2, 3, ..., n )

6.  **Mean and Variance**

* Mean = ( \mu = np )
* Variance = ( \sigma^2 = np(1-p) )

### Two Real-Life Examples

####  Example 1: Coin Toss

* Toss a coin 5 times.
* Count number of heads.
* Each toss has two outcomes (Head/Tail).
* Probability of head remains 0.5.
  âœ” Follows Binomial Distribution.

#### Example 2: Exam Result

* Each question has:

  * Correct answer (success)

### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10          # number of trials
p = 0.4         # probability of success
size = 1000     # sample size

# Generate random sample
sample = np.random.binomial(n, p, size)

# Plot histogram
plt.figure()
plt.hist(sample, bins=range(n + 2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Histogram of Binomial Distribution (n=10, p=0.4)")
plt.show()
```

### Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import math

def poisson_cdf(k, lam):
    """
    Calculate the cumulative distribution function (CDF)
    of a Poisson distribution.

    Parameters:
    k : int -> point at which CDF is calculated
    lam : float -> mean (Î»)

    Returns:
    float -> CDF value P(X â‰¤ k)
    """

    cumulative_prob = 0

    for i in range(0, k + 1):
        prob = (math.exp(-lam) * lam**i) / math.factorial(i)
        cumulative_prob += prob

    return cumulative_prob
```

```python
result = poisson_cdf(k=3, lam=2)
print(result)
```

### Q7. How Binomial distribution different from Poisson distribution?

Ans:
### Difference Between

### Binomial distribution and Poisson distribution

Both are **discrete probability distributions**, but they are used in different situations.

#### 1. Basic Idea

* **Binomial Distribution** â†’ Counts number of successes in **fixed number of trials**
* **Poisson Distribution** â†’ Counts number of events in a **fixed time or space interval**


#### 2. Key Differences (Comparison Table)

| Feature            | Binomial Distribution      | Poisson Distribution      |
| ------------------ | -------------------------- | ------------------------- |
| Number of trials   | Fixed (n)                  | Not fixed                 |
| Outcomes per trial | Two (Success/Failure)      | Event occurs or not       |
| Probability        | Constant probability **p** | Average rate **Î»**        |
| Mean               | ( np )                     | ( \lambda )               |
| Variance           | ( np(1-p) )                | ( \lambda )               |
| Parameters         | n and p                    | Î» only                    |
| Used for           | Repeated experiments       | Rare events in time/space |

#### 3. Shape Comparison

##### Binomial Distribution

* Can be symmetric or skewed
* Depends on n and p

#####  Poisson Distribution

* Usually right-skewed
* Used for rare events

#### 4. Relationship Between Them

When:

* n is very large
* p is very small
* np = Î»

Then **Binomial distribution approximates Poisson distribution**.

#### 5. Simple Examples

##### Binomial Example

* Toss a coin 10 times.
* Count number of heads.

##### Poisson Example

* Count number of calls received in 1 hour.

### Summary

* Use **Binomial** when trials are fixed and independent.
* Use **Poisson** when counting events occurring randomly over time or space.

### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# Parameters
lam = 5        # mean of Poisson distribution
size = 1000    # sample size

# Generate random sample
sample = np.random.poisson(lam, size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

# Print results
print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```
