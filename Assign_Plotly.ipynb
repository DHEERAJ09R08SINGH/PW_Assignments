### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
# Import required libraries
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Display first few rows (optional)
print(titanic.head())

# Create scatter plot using Plotly Express
fig = px.scatter(
    titanic,
    x="age",
    y="fare",
    title="Scatter Plot of Age vs Fare (Titanic Dataset)",
    labels={"age": "Age", "fare": "Fare"},
)

# Show the plot
fig.show()
```

### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
# Import required library
import plotly.express as px

# Load the tips dataset from plotly
tips = px.data.tips()

# Display first few rows (optional)
print(tips.head())

# Create box plot
fig = px.box(
    tips,
    x="day",          # Categorical column
    y="total_bill",   # Numerical column
    title="Box Plot of Total Bill by Day (Tips Dataset)"
)

# Show the plot
fig.show()
```

### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
# Import required library
import plotly.express as px

# Load dataset
tips = px.data.tips()

# Create histogram
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",              # Color by day
    pattern_shape="smoker",   # Pattern by smoker
    title="Histogram of Total Bill by Sex (Colored by Day, Pattern by Smoker)",
    barmode="group"           # Grouped bars for better comparison
)

# Show plot
fig.show()
```

### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
# Import required library
import plotly.express as px

# Load iris dataset
iris = px.data.iris()

# Display first few rows (optional)
print(iris.head())

# Create scatter matrix plot
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width",
                "petal_length", "petal_width"],
    color="species",   # Color by species
    title="Scatter Matrix Plot of Iris Dataset"
)

# Show plot
fig.show()
```

### Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans:

A Distplot (Distribution Plot) is used to visualize the distribution of a numerical variable.

It shows:

1)Frequency of values (like histogram)

2)Shape of distribution (normal, skewed, etc.)

3)Density curve (optional)

4)Helps understand spread, central tendency, and outliers

Earlier, Seaborn had distplot(), but in Plotly Express, we use:

->px.histogram() (for histogram)

->marginal="rug" or marginal="box" for extra distribution view

```python
# Import library
import plotly.express as px

# Load dataset
tips = px.data.tips()

# Create distribution plot (Histogram + density curve)
fig = px.histogram(
    tips,
    x="total_bill",
    nbins=30,
    marginal="rug",      # Adds rug plot
    histnorm="density",  # Normalize to show density
    title="Distribution Plot of Total Bill"
)

fig.show()
```
