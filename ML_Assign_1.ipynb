### Q1: Explain the following with an example:
1) Artificial Intelligence
2) Machine Learning
3) Deep Learning

Ans:

### 1. **Artificial Intelligence (AI)**

Artificial Intelligence is the field of computer science that focuses on creating machines or systems that can perform tasks that normally require human intelligence, such as reasoning, learning, decision-making, and problem solving.

**Example:**
A **voice assistant** like Google Assistant or Siri can understand your voice commands and answer questions or perform tasks.

---

### 2. **Machine Learning (ML)**

Machine Learning is a **subset of Artificial Intelligence** where computers learn patterns from data and improve their performance without being explicitly programmed for every task.

**Example:**
Email services like Gmail use machine learning to **detect spam emails** by learning patterns from millions of messages.

---

### 3. **Deep Learning (DL)**

Deep Learning is a **subset of Machine Learning** that uses neural networks with many layers (called deep neural networks) to learn complex patterns from large amounts of data.

**Example:**
Image recognition systems used in platforms like Google Photos can automatically **recognize faces, animals, and objects in photos** using deep learning.

---

✅ **Relationship between them:**

**Artificial Intelligence ⟶ Machine Learning ⟶ Deep Learning**

(AI is the broad field, ML is a subset of AI, and DL is a subset of ML.)


### Q2: What is supervised learning? List some examples of supervised learning.
Ans:

### **Supervised Learning**

**Supervised learning** is a type of **Machine Learning** where a model is trained using **labeled data**.
This means the input data is provided along with the **correct output (label)**, and the algorithm learns the relationship between them to make predictions on new data.

**Simple idea:**

Input + Correct Output → Model learns pattern → Predicts future outputs.

---

### **Example**

Suppose we want to predict whether an email is **Spam or Not Spam**.

| Email Words        | Label    |
| ------------------ | -------- |
| “Win money now”    | Spam     |
| “Meeting at 3 PM”  | Not Spam |
| “Claim your prize” | Spam     |

The model learns from these labeled examples and later classifies new emails.

---

### **Examples of Supervised Learning Applications**

1. **Spam Email Detection** – Classifying emails as spam or not spam.
2. **House Price Prediction** – Predicting house prices based on size, location, etc.
3. **Image Classification** – Identifying objects in images (cat, dog, car).
4. **Medical Diagnosis** – Predicting diseases based on patient data.
5. **Handwritten Digit Recognition** – Recognizing numbers written by humans.

---

### **Common Algorithms Used**

* Linear Regression
* Logistic Regression
* Decision Trees
* Support Vector Machines (SVM)
* Neural Networks

---

✅ **In short:**
Supervised learning uses **labeled training data** to learn patterns and **predict outputs for new inputs**.


### Q3: What is unsupervised learning? List some examples of unsupervised learning.
Ans:

### **Unsupervised Learning**

**Unsupervised learning** is a type of **Machine Learning** in which the model is trained using **unlabeled data**.
The algorithm tries to **discover hidden patterns or structures in the data** without being given the correct output.

**Simple idea:**

Input data → Model finds patterns or groups automatically.

---

### **Example**

Suppose a company has customer purchase data but **no predefined categories**.
The algorithm analyzes the data and groups customers into different clusters based on similar buying behavior.

Example groups:

* Frequent buyers
* Occasional buyers
* Discount-focused buyers

---

### **Examples of Unsupervised Learning Applications**

1. **Customer Segmentation** – Grouping customers based on purchasing behavior.
2. **Market Basket Analysis** – Finding products that are often bought together.
3. **Anomaly Detection** – Detecting unusual transactions in banking.
4. **Document Clustering** – Grouping similar articles or news topics.
5. **Image Segmentation** – Dividing images into meaningful parts.

---

### **Common Algorithms Used**

* K-Means Clustering
* Hierarchical Clustering
* Principal Component Analysis (PCA)
* Association Rule Learning

---

✅ **In short:**

Unsupervised learning analyzes **unlabeled data** and **automatically discovers hidden patterns, relationships, or groups** within the data.

### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

Ans:

#### **Differences between Supervised, Unsupervised, and Semi-Supervised Learning**

| Learning Type                | Description                                                                                    | Data Used                             | Goal                                          | Example                                      |
| ---------------------------- | ---------------------------------------------------------------------------------------------- | ------------------------------------- | --------------------------------------------- | -------------------------------------------- |
| **Supervised Learning**      | The model learns from **labeled data** where the correct output is already known.              | Labeled data (input + correct output) | Predict outcomes or classify data             | Email spam detection                         |
| **Unsupervised Learning**    | The model learns from **unlabeled data** and finds hidden patterns or groups in the data.      | Unlabeled data                        | Discover structure or clusters                | Customer segmentation                        |
| **Semi-Supervised Learning** | The model learns from **a small amount of labeled data and a large amount of unlabeled data**. | Both labeled and unlabeled data       | Improve learning when labeled data is limited | Image classification with few labeled images |

---

### **Key Points**

* **Supervised Learning**: Uses labeled datasets to train models for prediction.
* **Unsupervised Learning**: Identifies hidden patterns in unlabeled data.
* **Semi-supervised Learning**: Combines both labeled and unlabeled data for training.

---

✅ **Simple Summary**

* **Supervised Learning:** Learn with **answers given**.
* **Unsupervised Learning:** Learn **without answers**.
* **Semi-Supervised Learning:** Learn with **few answers and many unknowns**.


### Q6: What is train, test and validation split? Explain the importance of each term.

Ans:

### **Train, Test, and Validation Split**

In **Machine Learning**, a dataset is usually divided into **three parts**: **training set, validation set, and test set**. This helps evaluate how well a model learns and performs on new data.

---

### **1. Training Set**

The **training set** is the portion of data used to **train the machine learning model**.
The algorithm learns patterns and relationships from this data.

**Example:**
If we are predicting house prices, the training data may include house size, location, and the actual price. The model learns how these features relate to price.

**Importance:**

* Used to **build and train the model**
* Helps the algorithm learn patterns in the data
* Usually contains the **largest portion of the dataset**

---

### **2. Validation Set**

The **validation set** is used to **tune the model and adjust parameters** during training.

**Example:**
If we change parameters like the number of trees in a decision tree model, the validation set helps determine which setting gives the best performance.

**Importance:**

* Helps **optimize model parameters**
* Prevents **overfitting**
* Used during the training process to check model performance

---

### **3. Test Set**

The **test set** is used to **evaluate the final performance of the trained model**.

**Example:**
After the model is completely trained and tuned, we test it with new unseen data to see how accurately it predicts house prices.

**Importance:**

* Provides an **unbiased evaluation**
* Shows how the model performs on **new data**
* Used **only after training is finished**

---

### **Typical Data Split Example**

A common split of the dataset is:

* **Training set:** 70%
* **Validation set:** 15%
* **Test set:** 15%

---

✅ **Simple Summary**

* **Training set:** Used to **teach the model**
* **Validation set:** Used to **improve and tune the model**
* **Test set:** Used to **check final performance**


### Q7: How can unsupervised learning be used in anomaly detection?

Ans:

### **Unsupervised Learning in Anomaly Detection**

**Unsupervised Learning** can be used in **Anomaly Detection** to identify **unusual or abnormal patterns** in data when labeled examples of anomalies are not available.

In this approach, the algorithm first learns the **normal patterns or structure of the data**. Any data point that **does not fit these patterns** is considered an anomaly (outlier).

---

### **How It Works**

1. The model is trained on **unlabeled data**.
2. It identifies **clusters or patterns of normal behavior**.
3. Data points that **do not belong to any cluster or are far from normal patterns** are flagged as anomalies.

---

### **Example**

**Credit Card Fraud Detection**

Most credit card transactions follow normal spending patterns.
Using unsupervised learning, the system groups normal transactions together. If a transaction is **very different** (e.g., unusual location or large amount), it is flagged as **potential fraud**.

---

### **Common Algorithms Used**

* **K-Means Clustering**
* **Isolation Forest**
* **DBSCAN**
* **Principal Component Analysis (PCA)**

---

### **Applications of Anomaly Detection**

* Fraud detection in banking
* Network intrusion detection in cybersecurity
* Fault detection in machines
* Medical abnormality detection

---

✅ **In simple terms:**
Unsupervised learning detects anomalies by **learning normal patterns in data and identifying data points that significantly deviate from those patterns**.


### Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.

Ans:

#### **Commonly Used Supervised and Unsupervised Learning Algorithms**

In **Machine Learning**, different algorithms are used depending on whether the data is **labeled or unlabeled**.

---

### **1. Supervised Learning Algorithms**

These algorithms are used in **Supervised Learning**, where the model learns from **labeled data**.

**Common Algorithms:**

1. **Linear Regression** – Used for predicting continuous values (e.g., house price prediction).
2. **Logistic Regression** – Used for classification problems (e.g., spam detection).
3. **Decision Trees** – Tree-based model used for classification and regression.
4. **Support Vector Machine (SVM)** – Used for classification and pattern recognition.
5. **K-Nearest Neighbors (KNN)** – Classifies data based on nearest neighbors.
6. **Random Forest** – Ensemble method using multiple decision trees.
7. **Neural Networks** – Used for complex tasks like image and speech recognition.

---

### **2. Unsupervised Learning Algorithms**

These algorithms are used in **Unsupervised Learning**, where the model learns from **unlabeled data**.

**Common Algorithms:**

1. **K-Means Clustering** – Groups similar data points into clusters.
2. **Hierarchical Clustering** – Creates clusters in a hierarchical structure.
3. **DBSCAN** – Density-based clustering algorithm.
4. **Principal Component Analysis (PCA)** – Used for dimensionality reduction.
5. **Apriori Algorithm** – Used for association rule learning (market basket analysis).

---

✅ **Simple Summary**

| Learning Type             | Example Algorithms                                                              |
| ------------------------- | ------------------------------------------------------------------------------- |
| **Supervised Learning**   | Linear Regression, Logistic Regression, Decision Trees, SVM, KNN, Random Forest |
| **Unsupervised Learning** | K-Means, Hierarchical Clustering, DBSCAN, PCA, Apriori                          |
