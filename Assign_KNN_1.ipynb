### Q1. What is the KNN algorithm?

Ans: **K-Nearest Neighbors (KNN)** is a **supervised learning algorithm** in Machine Learning used for classification and regression.
It predicts the output of a new data point by finding the **k nearest labeled data points** and taking the majority vote (classification) or average (regression).


### Q2. How do you choose the value of K in KNN?

Ans: The value of **K** in K-Nearest Neighbors (KNN) is chosen based on model performance in Machine Learning.
A small K may cause overfitting, while a large K may cause underfitting, so the optimal K is selected using methods like cross-validation to achieve the best accuracy.


### Q3. What is the difference between KNN classifier and KNN regressor?

Ans:

**KNN Classifier vs KNN Regressor** in Machine Learning:

* **KNN Classifier:** Used for classification; it assigns a **class label** based on the majority vote of the k nearest neighbors.
* **KNN Regressor:** Used for regression; it predicts a **continuous value** by taking the average (or weighted average) of the k nearest neighbors.


### Q4. How do you measure the performance of KNN?

Ans:

The performance of **KNN** in Machine Learning is measured using different evaluation metrics depending on the task:

* **For Classification:** Accuracy, Precision, Recall, and F1-score are used.
* **For Regression:** Mean Squared Error (MSE), Mean Absolute Error (MAE), and R² score are used.


### Q5. What is the curse of dimensionality in KNN?

Ans:

The **curse of dimensionality** in KNN refers to the problem in Machine Learning where, as the number of features (dimensions) increases, the distance between data points becomes less meaningful.

As a result, KNN’s performance decreases because all points appear similarly distant, making it harder to find true nearest neighbors.


### Q6. How do you handle missing values in KNN?

Ans:

In **KNN** (in Machine Learning), missing values are handled before applying the algorithm:

* **Imputation:** Replace missing values using mean, median, or mode of the feature.
* **KNN Imputation:** Fill missing values using values from nearest neighbors.
* **Remove data:** Delete rows or columns if missing values are too many.


### Q7. Compare and contrast the performance of the KNN classifier and regressor. Which one is better for which type of problem?

Ans:

In Machine Learning, both KNN classifier and KNN regressor use nearest neighbors, but their performance differs based on the problem type:

* **KNN Classifier:** Performs well for **categorical (discrete) outputs**. It is better for problems like classification (e.g., spam detection) where results are class labels.
* **KNN Regressor:** Performs well for **continuous numerical outputs**. It is suitable for regression tasks (e.g., price prediction) where results are real values.

👉 **Conclusion:**

* Use **KNN Classifier** for classification problems.
* Use **KNN Regressor** for regression problems.


### Q8. What are the strengths and weaknesses of the KNN algorithm for classification and regression tasks, and how can these be addressed?

Ans:

In Machine Learning, the **KNN algorithm** has the following strengths and weaknesses:

### ✅ Strengths

* **Simple and easy to implement** (no training phase).
* **Works for both classification and regression**.
* **No assumption about data distribution** (non-parametric).

### ❌ Weaknesses

* **High prediction time** (computes distance with all points).
* **Sensitive to irrelevant features and noise**.
* **Curse of dimensionality** reduces performance in high dimensions.
* **Requires proper choice of K** and distance metric.

### 🔧 How to Address Them

* Use **KD-Tree / Ball Tree** to reduce computation time.
* Apply **feature scaling and selection** to remove irrelevant features.
* Use **dimensionality reduction** (e.g., PCA) to handle high dimensions.
* Choose optimal **K using cross-validation**.

👉 This improves efficiency, accuracy, and overall performance of KNN.


### Q9. What is the difference between Euclidean distance and Manhattan distance in KNN?

Ans:

In Machine Learning, the difference between Euclidean and Manhattan distance in KNN is:

* **Euclidean Distance:** Measures the **straight-line distance** between two points (shortest path).
* **Manhattan Distance:** Measures the **sum of absolute differences** along each dimension (grid-like path).

👉 **Key Difference:**
Euclidean is suitable for continuous, geometric space, while Manhattan works better when movement is restricted to axes or when features are independent.


### Q10. What is the role of feature scaling in KNN?

Ans:

In Machine Learning, **feature scaling** is important in KNN because the algorithm is based on distance calculations.

If features are on different scales, larger-valued features dominate the distance, so scaling (like normalization or standardization) ensures **all features contribute equally**, improving accuracy.
