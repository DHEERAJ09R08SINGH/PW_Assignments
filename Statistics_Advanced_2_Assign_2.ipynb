### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import scipy.stats as stats
import math

# Given values
mean = 50
std_dev = 5
n = 30  # assumed sample size
confidence = 0.95

# Standard error
se = std_dev / math.sqrt(n)

# t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin_of_error = t_critical * se

# Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("95% Confidence Interval:", (lower_bound, upper_bound))
```

### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

Ans:

### Interpretation

-> If the p-value is greater than 0.05, we conclude:

* There is not enough evidence to say the observed distribution is different from the expected distribution.

-> If the p-value is less than 0.05, we conclude:

* The observed color distribution significantly differs from the expected percentages.

```python
import numpy as np
from scipy.stats import chisquare

# Observed counts
observed = np.array([18, 22, 19, 12, 9, 20])

# Expected proportions
expected_proportions = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# Total sample size
total = observed.sum()

# Expected counts
expected = expected_proportions * total

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following

data:

   * Group A, Group B

Outcome 1 20 15

Outcome 2 10 25

Outcome 3 15 20

Interpret the results of the test.

Ans:

Step 1: Hypotheses

* H₀ (Null Hypothesis): Outcome and Group are independent.

* H₁ (Alternative Hypothesis): Outcome and Group are associated (dependent).

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create contingency table
data = np.array([
    [20, 15],
    [10, 25],
    [15, 20]
])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("Expected Frequencies:\n", expected)

# Decision at alpha = 0.05
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

#### Interpretation (α = 0.05)

* Since p-value (0.054) > 0.05,
➝ We fail to reject the null hypothesis.

#### Conclusion:

There is not enough evidence at the 5% significance level to conclude that the outcome is associated with the group.

In simple terms:

* The differences between Group A and Group B across outcomes are not statistically significant at the 0.05 level.

### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

Ans:

### Interpretation

* We are 95% confident that the true population proportion of smokers lies between:
9.16% and 14.84%
	​

This means that if many samples of 500 individuals were taken, about 95% of the calculated confidence intervals would contain the true population smoking proportion.

```python
import math
from scipy.stats import norm

# Given values
n = 500
x = 60

# Sample proportion
p_hat = x / n

# Z value for 95% confidence
z = norm.ppf(0.975)

# Standard error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Margin of error
margin = z * se

# Confidence interval
lower = p_hat - margin
upper = p_hat + margin

print("Sample Proportion:", p_hat)
print("95% Confidence Interval:", (lower, upper))
```

### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import math
from scipy.stats import t

# Given values
mean = 75
std_dev = 12
n = 36  # assumed sample size
confidence = 0.90

# Standard error
se = std_dev / math.sqrt(n)

# t critical value
t_critical = t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin = t_critical * se

# Confidence interval
lower = mean - margin
upper = mean + margin

print("90% Confidence Interval:", (lower, upper))
```

### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
# Chi-Square Distribution Plot (df = 10) and P-value Calculation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters
df = 10
chi_stat = 15

# Generate x values
x = np.linspace(0, 40, 500)
y = chi2.pdf(x, df)

# Plot the distribution
plt.figure()
plt.plot(x, y)
plt.fill_between(x, y, where=(x >= chi_stat))
plt.xlabel("Chi-Square Value")
plt.ylabel("Probability Density")
plt.title("Chi-Square Distribution (df = 10) with Shaded Area (χ² ≥ 15)")
plt.show()

# Calculate p-value (right-tail probability)
p_value = chi2.sf(chi_stat, df)

print("Chi-square Statistic:", chi_stat)
print("Degrees of Freedom:", df)
print("P-value:", p_value)
```

### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

Ans:

We are given:

* Sample size ( n = 1000 )
* Number who prefer Coke ( x = 520 )

---

###->  Step 1: Sample Proportion

P bar = x/n = 520/1000 = 0.52

So, **52%** of the sample prefer Coke.

###->  Step 2: 99% Confidence Interval Formula (Proportion)

CI = P bar ± Z_(α/2).sqrt[(P bar(1-p bar)/n]

For 99% confidence level:

* Z_{0.005} = 2.576

---

###-> Calculation

Standard Error:

* SE = sqrt[0.52 X 0.48 / 1000] ≈ 0.0158

Margin of Error:

* 2.576 X 0.0158 ≈ 0.0407

Confidence Interval:

* 0.52 ± 0.0407
* (0.4793, 0.5607)

---

###-> Interpretation

We are **99% confident** that the true population proportion of people who prefer Coke lies between:
* 47.93% and 56.07%

This means that if many samples of 1000 people were taken, about **99% of the calculated intervals would contain the true population proportion**.


```python
import math
from scipy.stats import norm

# Given values
n = 1000
x = 520

# Sample proportion
p_hat = x / n

# Z value for 99% confidence
z = norm.ppf(0.995)

# Standard error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Margin of error
margin = z * se

# Confidence interval
lower = p_hat - margin
upper = p_hat + margin

print("Sample Proportion:", p_hat)
print("99% Confidence Interval:", (lower, upper))
```

### Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

Ans: We will perform a **Chi-Square Goodness of Fit Test**.

---

###-> Step 1: Given Data

Total flips = 100

Observed:

* Tails = 45
* Heads = 55

Expected for a fair coin (50% each):

* Expected Tails = 50
* Expected Heads = 50

###-> Step 2: Hypotheses

* **H₀ (Null Hypothesis):** The coin is fair (p = 0.5).
* **H₁ (Alternative Hypothesis):** The coin is biased toward tails.

Significance level:
* α = 0.05

###-> Step-3 Manual Calculation

chi^2 = sum ((O - E)^2/E)

    = (45 - 50)^2 / 50 + (55 - 50)^2 / 50

    = 25/50 + 25/50

    = 0.5 + 0.5 = 1.0

Degrees of freedom:

* df = k - 1 = 2 - 1 = 1

P-value ≈ **0.317**

---

###-> Interpretation

Since:

p-value(0.317) > 0.05


We **fail to reject the null hypothesis**.

###-> Conclusion:

There is **no sufficient evidence** at the 5% significance level to conclude that the coin is biased toward tails.

The observed difference (45 vs 50) is likely due to random variation.


```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([45, 55])

# Expected frequencies (fair coin)
expected = np.array([50, 50])

# Perform chi-square goodness of fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.

#### smoker/non-smoker, Lung Cancer: Yes, Lung Cancer: No
Smoker 60 140
Non-smoker 30 170

Use a significance level of 0.05.

## Ans:


We will conduct a **Chi-Square Test for Independence**.

---

###-> Step 1: Given Contingency Table

| Smoking Status   | Lung Cancer: Yes | Lung Cancer: No | Row Total |
| ---------------- | ---------------- | --------------- | --------- |
| Smoker           | 60               | 140             | 200       |
| Non-smoker       | 30               | 170             | 200       |
| **Column Total** | **90**           | **310**         | **400**   |

Total sample size = **400**

###-> Step 2: Hypotheses

* **H₀ (Null Hypothesis):** Smoking status and lung cancer diagnosis are independent.
* **H₁ (Alternative Hypothesis):** There is an association between smoking and lung cancer.

Significance level:
* α = 0.05

###-> Step 3: Expected Frequencies

E = (Row Total)(Column Total)/(Grand Total)

For Smoker & Yes:

* E = (200 X 90) /(400) = 45

For Smoker & No:

* E = (200 X 310)/(400) = 155


For Non-smoker & Yes:

* E = 45

For Non-smoker & No:

* E = 155


###-> Step 4: Chi-Square Statistic

chi^2 = sum [(O - E)^2/E]

      = (60-45)^2/45 + (140-155)^2/155 + (30-45)^2/45 + (170-155)^2/155
      = 5 + 1.45 + 5 + 1.45
      
chi^2 ≈ 12.90

Degrees of freedom:

* df = (r-1)(c-1) = (2-1)(2-1) = 1

###-> Step 5: Critical Value / P-value

For:

* df = 1
* α = 0.05

Critical value from chi-square table:

* chi^2_(0.05,1) = 3.84

Since:

* 12.90 > 3.84

OR equivalently,

P-value < 0.001

---

### Final Conclusion

Because:

* chi^2 = 12.90 and p < 0.05

We **reject the null hypothesis**.

###  Interpretation:

There is a **statistically significant association** between smoking status and lung cancer diagnosis at the 5% significance level.

Smokers appear to have a higher rate of lung cancer compared to non-smokers.


### Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin.


#### Milk Chocolate, Dark Chocolate, White Chocolate
U.S. (n=500) 200 150 150
U.K. (n=500) 225 175 100

Use a significance level of 0.01.

Ans:

We will conduct a **Chi-Square Test for Independence**.

---

###-> Step 1: Given Contingency Table

| Country          | Milk    | Dark    | White   | Row Total |
| ---------------- | ------- | ------- | ------- | --------- |
| U.S.             | 200     | 150     | 150     | 500       |
| U.K.             | 225     | 175     | 100     | 500       |
| **Column Total** | **425** | **325** | **250** | **1000**  |


###-> Step 2: Hypotheses

* **H₀:** Chocolate preference and country are independent.
* **H₁:** There is an association between chocolate preference and country.

Significance level:
* α = 0.01



###-> Step-3: Manual Calculation

Expected frequencies (for each country):

* Milk:
   * E = 500 X 425 / 1000 = 212.5

* Dark:
   * E = 500 X 325 / 1000 = 162.5

* White:
  * E = 500 X 250 / 1000 = 125

Chi-square statistic:
* chi^2 ≈ 14.00

Degrees of freedom:
* df = (2-1)(3-1) = 2


Critical value at ( α = 0.01 ), df = 2:
* chi^2_(0.01,2) = 9.21

Since:
* 14.00 > 9.21

P-value ≈ **0.0009**

---

### Final Conclusion

Because:

* p < 0.01


We **reject the null hypothesis**.

### Interpretation:

There is a **statistically significant association** between chocolate preference and country of origin at the 1% significance level.

Chocolate preferences differ between the U.S. and the U.K.






```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
data = np.array([
    [200, 150, 150],   # U.S.
    [225, 175, 100]    # U.K.
])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)

print("\nExpected Frequencies:")
print(expected)

# Decision at alpha = 0.01
alpha = 0.01
if p < alpha:
    print("\nReject the null hypothesis.")
    print("There is a significant association between chocolate preference and country.")
else:
    print("\nFail to reject the null hypothesis.")
    print("There is no significant association between chocolate preference and country.")
```

### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use asignificance level of 0.05.

Ans:

We will conduct a **one-sample t-test** (since population standard deviation is unknown and n = 30).

---

### Step 1: Given Data

* Sample size ( n = 30 )
* Sample mean ( x bar = 72 )
* Sample standard deviation ( s = 10 )
* Hypothesized mean ( μ0​ = 70 )
* Significance level ( α = 0.05 )

### Step 2: Hypotheses

Since the question says *“significantly different”*, this is a **two-tailed test**:

* H_0: μ = 70

* H_1: μ != 70

### Step 3: Test Statistic

* t = (x bar -  μ0)/ (s /sqrt(n))

Standard Error:

* SE = 10/sqrt(30) ≈ 1.826

t-statistic:

* t = (72 - 70)/1.826

* t ≈ 1.095

Degrees of freedom:
* df = n - 1 = 29


### Step-4 Critical Value Method

For:

* df = 29
* α = 0.05  (two-tailed)

Critical value:

* t_(0.025,29) ≈ ± 2.045

Since:
* 1.095 < 2.045

We do **not** reject (H_0).

---

### Final Conclusion

P-value ≈ **0.28**

Because:

[
p > 0.05
]

We **fail to reject the null hypothesis**.

###  Interpretation:

There is **not enough evidence** at the 5% significance level to conclude that the population mean is different from 70.

The sample mean of 72 is **not statistically significantly different** from 70.


```python
import math
from scipy.stats import t

# Given values
n = 30
mean = 72
std_dev = 10
mu_0 = 70
alpha = 0.05

# Standard error
se = std_dev / math.sqrt(n)

# t statistic
t_stat = (mean - mu_0) / se

# Degrees of freedom
df = n - 1

# Two-tailed p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

print("t-statistic:", t_stat)
print("Degrees of freedom:", df)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
