### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

Ans:

### **1. Key Differences Between t-test and z-test**

| Feature                           | **t-test**                                              | **z-test**                                                          |
| --------------------------------- | ------------------------------------------------------- | ------------------------------------------------------------------- |
| **Population Standard Deviation** | Unknown (use sample standard deviation `s`)             | Known (population standard deviation `σ`)                           |
| **Sample Size**                   | Small (typically n < 30)                                | Large (typically n ≥ 30)                                            |
| **Distribution Used**             | Student’s t-distribution                                | Standard normal distribution (z-distribution)                       |
| **Purpose**                       | Test hypotheses about population mean when σ is unknown | Test hypotheses about population mean when σ is known or n is large |

---

### **2. When to Use Each Test**

#### **t-test Example**

**Scenario:** we want to test whether a new teaching method improves students’ test scores.

* we randomly sample **25 students** and calculate their average score.
* Population standard deviation **σ is unknown**.

Here, we use a **t-test** because the sample is small and σ is unknown.

---

#### **z-test Example**

**Scenario:** A factory claims that the average weight of a packaged cereal box is 500g. You take a **random sample of 100 boxes**. we **know from past data that the population standard deviation σ = 10g**.

* we want to test whether the actual average weight is different from 500g.

Here, we use a **z-test** because the sample size is large and σ is known.

---
### **Summary:**

* Use **t-test** for small samples and unknown population SD.
* Use **z-test** for large samples or when population SD is known.


### Q2: Differentiate between one-tailed and two-tailed tests.

Ans:

### **1. Definition**

| Feature        | **One-Tailed Test**                                                                    | **Two-Tailed Test**                                                        |
| -------------- | -------------------------------------------------------------------------------------- | -------------------------------------------------------------------------- |
| **Purpose**    | Tests for a deviation in **one specific direction** (either greater than or less than) | Tests for a deviation in **both directions** (either greater or less than) |
| **Hypotheses** | - H₀: μ = μ₀                                                                           |                                                                            |

* H₁: μ > μ₀ (right-tailed) or μ < μ₀ (left-tailed) | - H₀: μ = μ₀
* H₁: μ ≠ μ₀ |
  | **Critical Region** | Only **one tail** of the distribution | **Both tails** of the distribution |
  | **Significance Level (α)** | All α is in **one tail** | α is **split between two tails** (α/2 in each tail) |

---

### **2. When to Use**

#### **One-Tailed Test**

* Use when the research hypothesis predicts a **specific direction**.
  **Example:**
  A company claims that a new fertilizer **increases** crop yield. we test if the mean yield is **greater than** the standard yield.

#### **Two-Tailed Test**

* Use when we only care if there is **any difference**, without specifying the direction.
  **Example:**
  we want to check whether a new teaching method changes test scores, without assuming it improves or worsens them.

---

### **3. Visual Representation**

* **One-tailed (right-tail example):**

```
α -------->
|         |
0       Critical value
```

* **Two-tailed:**

```
<---- α/2        α/2 ---->
|                 |
Critical values at both ends
```

---

**Key Tip:**

* If you have a **specific directional claim**, use **one-tailed**.
* If you just want to see **any difference**, use **two-tailed**.



### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Ans:

### **1. Concept**

| Error Type                  | Definition                                                                 | Symbol | Consequence                                                       |
| --------------------------- | -------------------------------------------------------------------------- | ------ | ----------------------------------------------------------------- |
| **Type I Error (α error)**  | Rejecting the null hypothesis **H₀** when it is actually **true**          | α      | False positive — thinking there is an effect when there isn’t one |
| **Type II Error (β error)** | Failing to reject the null hypothesis **H₀** when it is actually **false** | β      | False negative — failing to detect an effect that exists          |

---

### **2. Examples**

#### **Type I Error Example**

**Scenario:** A new drug is tested to see if it lowers blood pressure.

* **H₀:** The drug has no effect.
* **H₁:** The drug lowers blood pressure.

**Type I error:** Concluding that the drug works (rejecting H₀) **when in reality it doesn’t**.

* Consequence: People may take an ineffective drug.

---

#### **Type II Error Example**

**Scenario:** Same drug test.

* **H₀:** The drug has no effect.
* **H₁:** The drug lowers blood pressure.

**Type II error:** Concluding that the drug does **not work** (failing to reject H₀) **when in reality it does**.

* Consequence: A beneficial drug may be discarded.

---

### **3. Key Points to Remember**

* **Type I error** → False alarm (thinking effect exists when it doesn’t).
* **Type II error** → Missed detection (not noticing a real effect).
* Reducing **α** (Type I error) increases **β** (Type II error), and vice versa — there’s a trade-off.



### Q4: Explain Bayes's theorem with an example.
Ans:
### **1. Concept**

**Bayes’s Theorem** is a way to **update the probability of an event** based on new evidence. It connects **conditional probabilities**:

P(A|B) = [P(B|A) . P(A)]/P(B)

Where:

* P(A|B) = Probability of event A given that B has occurred (**posterior probability**)
* P(B|A) = Probability of event B given that A has occurred (**likelihood**)
* P(A) = Probability of event A (**prior probability**)
* P(B) = Probability of event B (**total probability of B**)

---

### **2. Example**

**Scenario:** A medical test for a disease.

* 1% of the population has the disease: (P(D) = 0.01)
* Test correctly detects disease 99% of the time: (P(\text{Positive}|D) = 0.99)
* Test falsely gives positive for healthy people 5% of the time: (P(\text{Positive}|\text{No Disease}) = 0.05)

**Question:** If someone tests positive, what is the probability they actually have the disease? P(D|Positive)

**Step 1: Total probability of testing positive (P(Positive)):**

P(Positive) = P(Positive|D) . P(D) + P(Positive|No Disease) .P(No Disease)

= 0.99 . 0.01 + 0.05 . 0.99

= 0.0099 + 0.0495 = 0.0594

**Step 2: Apply Bayes’s Theorem:**

P(D∣Positive) = [P(Positive∣D) . P(D)]/
           P(Positive)

= 0.99 . 0.01 / 0.0594
= 0.0099/0.0594 ≈ 0.1667

 **Answer:** About **16.7% chance** the person actually has the disease, even after a positive test.

---

### **3. Key Insight**

Bayes’s Theorem shows that even a highly accurate test can have a low probability of a true positive when the **prior probability (prevalence) is low**.



### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Ans:

A **confidence interval (CI)** is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as the population mean).

It gives an **estimate + margin of error** instead of just a single value.

---

### **1. General Formula**

For population mean:

#### When population standard deviation (σ) is known → **Z-interval**

CI = x bar ± [Z_(α/2) . σ/sqrt(n)]

#### When σ is unknown → **t-interval**

CI = x bar ± [t_(α/2) . σ/sqrt(n)]

Where:

* X bar = Sample mean
* Z_(α/2) = Z critical value
* t_(α/2) = t critical value
* σ = Population standard deviation
* s = Sample standard deviation
* n = Sample size

---

### **2. Example**

**Scenario:**
A sample of 36 students has an average score of 70.
Population standard deviation is known: σ = 12.
Find the **95% confidence interval**.

For 95% confidence level:
Z_(α/2) = 1.96

#### **Step 1: Calculate Standard Error**

 SE = σ/sqrt(n)
    = 12/sqrt(36) = 2

#### **Step 2: Calculate Margin of Error**

ME = 1.96 X 2 = 3.92

#### **Step 3: Construct Confidence Interval**

70±3.92

(66.08,\ 73.92)

---

### Final Answer:

We are **95% confident** that the true population mean lies between **66.08 and 73.92**.

---

### **3. Interpretation (Very Important for Exams)**

A 95% confidence interval means:

If we take many samples and construct intervals in the same way, about **95% of those intervals will contain the true population mean**.

It does NOT mean there is a 95% probability that the true mean is in this specific interval.

---

### 4. Key Points

* Larger sample size → Narrower interval
* Higher confidence level → Wider interval
* More variability → Wider interval



### Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

Ans:

### Sample Problem

A factory produces light bulbs from two machines:

* **Machine A** produces 60% of bulbs and has a 2% defect rate.
* **Machine B** produces 40% of bulbs and has a 5% defect rate.

A randomly selected bulb is found to be defective.

What is the probability that it was produced by **Machine B**?

---

#### **Step 1: Define Probabilities**

Let:

* (A) = Bulb from Machine A
* (B) = Bulb from Machine B
* (D) = Bulb is defective

Given:

P(A)=0.6

P(B)=0.4

P(D|A)=0.02

P(D|B)=0.05


We need to find:

P(B|D)

### **Step 2: Find Total Probability of Defective Bulb**

P(D) = P(D|A)P(A) + P(D|B)P(B)

P(D) = (0.02)(0.6) + (0.05)(0.4)

P(D) = 0.012 + 0.020 = 0.032

### **Step 3: Apply Bayes’ Theorem**

P(B|D) = P(D|B)P(B)/P(D)

P(B|D) = (0.05)(0.4)/0.032

P(B|D) = 0.020/0.032

P(B|D) = 0.625

---

### Final Answer:

The probability that a defective bulb came from **Machine B** is:

0.625 or 62.5%

---

### Interpretation

Even though Machine B produces fewer bulbs (40%), it has a higher defect rate. Therefore, when a defective bulb is found, it is **more likely** to have come from Machine B.



### Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.

Ans:

To calculate a **95% confidence interval**, we need:

* Sample mean (x bar) = 50
* Sample standard deviation (s) = 5
* Sample size (n) → ⚠️ *Not given*

Since the sample size is not provided, we will assume **n = 25** (a common exam assumption for demonstration).

---

### Step 1: Formula (σ unknown → use t or z)

For simplicity (assuming large sample or using z ≈ 1.96):

CI = x bar ± [Z_(α/2) . s/sqrt(n)]

For 95% confidence:

Z_(α/2) = 1.96

### Step 2: Calculate Standard Error

SE = 5/sqrt(25) = 1


### Step 3: Margin of Error

ME = 1.96 X 1

ME = 1.96

### Step 4: Confidence Interval

50±1.96

(48.04, 51.96)


---

### Final Answer:

The **95% confidence interval** is:

(48.04, 51.96)

---

#### Interpretation:

We are **95% confident** that the true population mean lies between **48.04 and 51.96**.

This does NOT mean there is a 95% probability that the true mean is in this interval.
It means that if we repeated the sampling process many times, about **95% of the constructed intervals would contain the true mean**.



````python
import math

# Given values
mean = 50
std_dev = 5
n = 25   # change this to your sample size

# Z value for 95% confidence
z = 1.96

# Standard Error
se = std_dev / math.sqrt(n)

# Margin of Error
margin_error = z * se

# Confidence Interval
lower = mean - margin_error
upper = mean + margin_error

print("95% Confidence Interval:", (lower, upper))
````

### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

Ans:

The **margin of error (MOE)** is the amount added to and subtracted from the sample estimate to create a confidence interval.

Confidence Interval=xˉ±Margin of Error

It represents how much uncertainty exists in the estimate.

---

#### **1. Formula**

For a mean (using Z distribution):

MOE=Z_(α/2)​⋅​σ​/sqrt(n)

Where:

* Z_(α/2) = Critical value (e.g., 1.96 for 95%)
* σ = Standard deviation
* n = Sample size

---

#### **2. How Sample Size Affects Margin of Error**

The key part of the formula is:

1/sqrt(n)

* As **sample size (n) increases**,
* The denominator (sqrt(n)) increases,
* So the margin of error **decreases**.

#### Important Relationship:

* Larger sample size → Smaller MOE → More precise estimate
* Smaller sample size → Larger MOE → Less precise estimate

---

#### **3. Numerical Example**

Assume:

* Standard deviation = 10
* 95% confidence level → (Z = 1.96)

#### Case 1: n = 25

MOE = 1.96 x 2 = 3.92

#### Case 2: n = 100

MOE = 1.96 x 1 = 1.96


---

#### Result:

When sample size increased from 25 to 100,
the margin of error decreased from **3.92 to 1.96** (cut in half).

---

#### **4. Real-Life Scenario**

Imagine conducting an election poll:

* If you survey **100 voters**, your margin of error might be ±5%.
* If you survey **10,000 voters**, your margin of error might drop to ±1%.

A larger sample gives a more accurate estimate of the true population opinion.

---

#### Final Summary

Margin of Error measures the **precision** of your estimate.
It decreases when:

* Sample size increases
* Variability decreases
* Confidence level decreases


````python
import math
import matplotlib.pyplot as plt

# Given values
sigma = 10
z = 1.96  # 95% confidence level

# Different sample sizes
sample_sizes = [10, 25, 50, 100, 200, 500, 1000]

# Calculate Margin of Error for each sample size
margin_of_error = []

for n in sample_sizes:
    moe = z * (sigma / math.sqrt(n))
    margin_of_error.append(moe)

# Print results
for n, moe in zip(sample_sizes, margin_of_error):
    print(f"Sample Size: {n}, Margin of Error: {moe:.4f}")

# Plot the relationship
plt.figure()
plt.plot(sample_sizes, margin_of_error)
plt.xlabel("Sample Size (n)")
plt.ylabel("Margin of Error")
plt.title("Effect of Sample Size on Margin of Error")
plt.show()
````

### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

####  Z-Score Calculation
Formula:
    z=(x−μ​)/σ

z=(75−70)/5

### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

Ans:

### Step 1: Define Hypotheses

We want to test whether the drug is **significantly effective**.

Assume:

* Population mean weight loss without drug = 0 pounds

### Hypotheses:

* **Null Hypothesis (H₀):** μ = 0
  (The drug has no effect)

* **Alternative Hypothesis (H₁):** μ > 0
  (The drug causes weight loss)

This is a **one-tailed t-test**.

### Step 2: Given Data

* Sample size (n) = 50
* Sample mean (x̄) = 6 pounds
* Sample standard deviation (s) = 2.5 pounds
* Confidence level = 95%
* Significance level α = 0.05


### Step 3: Test Statistic Formula (t-test)

t=(​xˉ−μ​)/[s/sqrt(n)]


### Step 4: Calculate t-value

#### Standard Error:

* SE = 2.5/sqrt(50) ≈ 0.354


#### t-value:

* t = (6 - 0)/(0.354)

* t ≈ 16.95

### Step 5: Critical Value

Degrees of freedom:
* df = n - 1 = 49
* For α = 0.05 (one-tailed) and df = 49:
* t_{critical} ≈ 1.676

### Step 6: Decision Rule

Since:

16.95 > 1.676

We **reject the null hypothesis**.

---

### Final Conclusion

At the **95% confidence level**, the drug is **statistically significantly effective**.

The weight loss is far greater than what would occur by random chance.




````python
import math
from scipy import stats

# Given values
mean = 6
mu = 0
s = 2.5
n = 50

# Standard Error
se = s / math.sqrt(n)

# t statistic
t_stat = (mean - mu) / se

# Degrees of freedom
df = n - 1

# Critical value (one-tailed, 95%)
t_critical = stats.t.ppf(0.95, df)

# p-value
p_value = 1 - stats.t.cdf(t_stat, df)

print("t-statistic:", t_stat)
print("Critical value:", t_critical)
print("p-value:", p_value)

if t_stat > t_critical:
    print("Reject the null hypothesis. Drug is significantly effective.")
else:
    print("Fail to reject the null hypothesis.")
````

### Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

Ans:

Interpretation

We are 95% confident that the true proportion of people satisfied with their job lies between 60.8% and 69.2%.

````python
import math

# Given values
p_hat = 0.65
n = 500
z = 1.96  # 95% confidence level

# Standard Error
se = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of Error
margin_error = z * se

# Confidence Interval
lower = p_hat - margin_error
upper = p_hat + margin_error

print("95% Confidence Interval:", (lower, upper))
print("In percentage:", (lower * 100, upper * 100))
````

### Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

Ans:

To properly conduct an independent two-sample t-test, we need the sample sizes for both groups.

Since they are not provided, I will assume:
* n_A = 30
* n_B = 30

### Step 1: Define Hypotheses

We are checking if there is a difference between the two teaching methods.

* H₀: μ₁ = μ₂ (No difference)
* H₁: μ₁ ≠ μ₂ (There is a difference)

This is a two-tailed t-test.

Significance level:
α=0.01

### Step 2: Given Data

| Group | Mean | SD | n  |
| ----- | ---- | -- | -- |
| A     | 85   | 6  | 30 |
| B     | 82   | 5  | 30 |

### Step 3: Test Statistic (Independent t-test)
* t

### Step 4: Compute Standard Error
* SE≈1.426

### Step 5: Calculate t-value

* t≈2.10

### Step 6: Critical Value
Degrees of freedom (approx):

* df=n_1+n_2−2=58

For:
* α = 0.01 (two-tailed)
* df = 58

Critical t ≈ ±2.66

### Step 7: Decision

* Since: 2.10 < 2.66

* We fail to reject the null  hypothesis.

````python
import math
from scipy import stats

# Given values
mean1 = 85
mean2 = 82
s1 = 6
s2 = 5
n1 = 30
n2 = 30

# Standard Error
se = math.sqrt((s1**2 / n1) + (s2**2 / n2))

# t-statistic
t_stat = (mean1 - mean2) / se

# Degrees of freedom
df = n1 + n2 - 2

# Critical value (two-tailed, alpha = 0.01)
t_critical = stats.t.ppf(1 - 0.01/2, df)

# p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print("t-statistic:", t_stat)
print("Critical value:", t_critical)
print("p-value:", p_value)

if abs(t_stat) > t_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
````

### Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.




````python
import math
from scipy import stats

# Given values
sample_mean = 65
sigma = 8
n = 50
confidence_level = 0.90

# Z critical value
z = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard Error
se = sigma / math.sqrt(n)

# Margin of Error
margin_error = z * se

# Confidence Interval
lower = sample_mean - margin_error
upper = sample_mean + margin_error

print("Z critical value:", round(z, 3))
print("Standard Error:", round(se, 3))
print("Margin of Error:", round(margin_error, 3))
print("90% Confidence Interval:", (round(lower, 2), round(upper, 2)))
````

### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

Ans:

### Conclusion

At the 90% confidence level, caffeine has a statistically significant effect on reaction time.

Because the sample mean (0.25 sec) is lower than the assumed normal time (0.30 sec), caffeine appears to reduce reaction time (faster response).

````python
import math
from scipy import stats

# Given values
mean_sample = 0.25
mu = 0.30  # assumed normal reaction time
s = 0.05
n = 30
alpha = 0.10

# Standard Error
se = s / math.sqrt(n)

# t-statistic
t_stat = (mean_sample - mu) / se

# Degrees of freedom
df = n - 1

# Critical value (two-tailed)
t_critical = stats.t.ppf(1 - alpha/2, df)

# p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print("t-statistic:", round(t_stat, 3))
print("Critical value:", round(t_critical, 3))
print("p-value:", round(p_value, 5))

if abs(t_stat) > t_critical:
    print("Reject the null hypothesis. Caffeine has a significant effect.")
else:
    print("Fail to reject the null hypothesis.")
````
