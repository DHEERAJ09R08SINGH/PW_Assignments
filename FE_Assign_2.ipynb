### Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

Ans:

### **Min–Max Scaling (Normalization):**

**Min–Max Scaling** is a **data preprocessing technique** used to **normalize numerical data** so that all values fall within a **fixed range**, usually **0 to 1**.

It rescales the original data without changing the relationships between the values.

#### **Formula:**

X_(scaled) = X - X_(min)/ X_(max) - X_(min)

Where:

* X = original value
* X_(min) = minimum value in the dataset
* X_(max) = maximum value in the dataset

The transformed value will lie between **0 and 1**.

---

### **Why Min–Max Scaling is Used:**

Min–Max scaling is useful because many machine learning algorithms perform better when features are on a **similar scale**.

It helps in algorithms such as:

* **KNN**
* **K-Means**
* **Neural Networks**
* **Gradient Descent based algorithms**

Without scaling, features with larger ranges can **dominate** those with smaller ranges.

---

### **Example**

Suppose we have the following feature representing **house sizes (sq ft):**

| House | Size |
| ----- | ---- |
| A     | 1000 |
| B     | 1500 |
| C     | 2000 |
| D     | 2500 |

Here

* **Min value = 1000**
* **Max value = 2500**

Apply Min–Max Scaling.

#### For House A

[
(1000-1000)/(2500-1000)=0
]

#### For House B

[
(1500-1000)/(2500-1000)=0.33
]

#### For House C

[
(2000-1000)/(2500-1000)=0.67
]

#### For House D

[
(2500-1000)/(2500-1000)=1
]

| House | Original Size | Scaled Value |
| ----- | ------------- | ------------ |
| A     | 1000          | 0.00         |
| B     | 1500          | 0.33         |
| C     | 2000          | 0.67         |
| D     | 2500          | 1.00         |

Now all values lie between **0 and 1**.

---

### **Conclusion**

Min–Max scaling **normalizes data by rescaling it to a fixed range (usually 0–1)**, ensuring that all features contribute equally to machine learning models and preventing features with larger magnitudes from dominating the learning process.



### Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling? Provide an example to illustrate its application.

Ans:

### **Unit Vector Technique (Normalization):**

The **Unit Vector technique** is a **feature scaling method** that transforms data so that the **length (magnitude) of each data vector becomes 1**.
It scales the feature values by dividing them by the **Euclidean norm (vector magnitude)** of the data.

This method is also called **Vector Normalization**.

#### **Formula:**

X_(normalized) = X / [sqrt(x_1^2 + x_2^2 + x_3^2 + ... + x_n^2)]

Where:

* X = original feature value
* Denominator = **Euclidean length (norm) of the vector**

After transformation, the **vector length becomes 1**.

---

## **Example:**

Suppose we have a data point with two features:

| Feature1 | Feature2 |
| -------- | -------- |
| 3        | 4        |

### Step 1: Calculate the vector magnitude

sqrt(3^2 + 4^2) = sqrt(9 + 16) = sqrt(25) = 5


### Step 2: Divide each value by the magnitude

Feature1 = 3/5 = 0.6

Feature2 = 4/5 = 0.8

### Normalized vector:

| Feature1 | Feature2 |
| -------- | -------- |
| 0.6      | 0.8      |

Now the vector length is:

sqrt(0.6^2 + 0.8^2) = 1

---

## **Difference Between Unit Vector Scaling and Min–Max Scaling:**

| Aspect       | **Unit Vector Scaling**                        | **Min–Max Scaling**                      |
| ------------ | ---------------------------------------------- | ---------------------------------------- |
| Purpose      | Normalizes vector length to **1**              | Scales values to a **fixed range (0–1)** |
| Formula      | Divide by **vector magnitude**                 | ((X - min)/(max - min))                  |
| Applied On   | Entire **data vector**                         | Each **feature independently**           |
| Result Range | Not fixed (depends on vector)                  | Always **0 to 1**                        |
| Common Use   | **Text mining, cosine similarity, clustering** | **General ML preprocessing**             |

---

### **Simple Interpretation**

* **Min–Max Scaling:** Rescales data so values fall **between 0 and 1**.
* **Unit Vector Scaling:** Rescales data so the **vector length becomes 1**.

---

### **Conclusion**

The **Unit Vector technique** normalizes data by dividing each feature value by the **magnitude of the vector**, ensuring that the **vector length equals 1**, while **Min–Max scaling rescales features to a fixed range (0–1)** based on minimum and maximum values.



### Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

Ans:
### **Principal Component Analysis (PCA):**

**Principal Component Analysis (PCA)** is a **dimensionality reduction technique** used in machine learning and statistics to **reduce the number of features (variables) in a dataset while retaining most of the important information (variance).**

It transforms the original correlated variables into a **new set of uncorrelated variables called Principal Components**.

These components are ordered such that:

* **PC1 (First Principal Component)** captures the **maximum variance**
* **PC2** captures the **second highest variance**
* and so on.

---

### **How PCA Works in Dimensionality Reduction:**

1. **Standardize the data** (mean = 0, variance = 1).
2. **Compute the covariance matrix** to understand relationships between variables.
3. **Calculate eigenvalues and eigenvectors**.
4. **Select principal components** with the highest eigenvalues.
5. **Transform the data** into the new feature space with fewer dimensions.

This reduces the dataset size while keeping most of the important information.

---

### **Example:**

Suppose a dataset contains the following features about houses:

* Size of house (sq ft)
* Number of rooms
* Total floor area
* Number of bathrooms

Some of these features are **highly correlated** (for example, house size and floor area).

Using **PCA**, these 4 features can be reduced to **2 principal components**:

| Original Features | After PCA |
| ----------------- | --------- |
| Size              | PC1       |
| Rooms             |           |
| Floor Area        | PC2       |
| Bathrooms         |           |

Now instead of **4 variables**, the dataset uses **2 principal components** that still represent most of the information.

---

### **Real-World Example**

In **image processing**:

* An image may contain **thousands of pixels (features)**.
* PCA reduces these features to **a smaller number of components** while preserving the main patterns of the image.

This helps in:

* **Faster model training**
* **Less storage**
* **Reduced noise**

---

### **Advantages of PCA**

* Reduces **computational cost**
* Removes **redundant features**
* Helps **visualize high-dimensional data**
* Improves **model performance**

---

### **Conclusion**

**PCA is a dimensionality reduction technique that transforms correlated features into a smaller number of uncorrelated principal components while preserving the maximum variance in the dataset.**


```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Sample dataset
data = {
    'Feature1': [2, 4, 6, 8],
    'Feature2': [1, 3, 5, 7],
    'Feature3': [10, 20, 30, 40]
}

df = pd.DataFrame(data)

# Step 1: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Step 3: Convert to DataFrame
pca_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])

print("Original Data:\n", df)
print("\nPCA Transformed Data:\n", pca_df)
```

### Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the values to a range of -1 to 1.

Ans:

### **Min–Max Scaling to Range [-1, 1]:**

To scale data to a range **[-1, 1]**, we use the generalized **Min–Max scaling formula**:

X_(scaled) = a + [(X - X_(min))(b-a) / X_(max)-X_(min)]

Where:

* a = -1 (new minimum)
* b = 1 (new maximum)
* X_(min) = minimum value in dataset
* X_(max) = maximum value in dataset

---

### **Step 1: Identify Min and Max:**

Dataset: **[1, 5, 10, 15, 20]**

* (X_{min} = 1)
* (X_{max} = 20)

Range difference:

[
X_{max} - X_{min} = 20 - 1 = 19
]

New range:

[
b - a = 1 - (-1) = 2
]

---

### **Step 2: Apply Formula**

#### For (X = 1)

[
-1 + (1-1)2 / 19 = -1
]

#### For (X = 5)

[
= -1 + (5-1)2 / 19
= -1 + (8)(19)
= -0.58
]

#### For (X = 10)

[
= -1 + (10-1)2 / 19
= -1 + 18 / 19
= -0.05
]

#### For (X = 15)

[
= -1 + (15-1)2 / 19
= -1 + 28 / 19
= 0.47
]

#### For (X = 20)

[
= -1 + (20-1)2 / 19
= -1 + 2
= 1
]

---

### **Final Scaled Values**

| Original Value | Scaled Value |
| -------------- | ------------ |
| 1              | -1.00        |
| 5              | -0.58        |
| 10             | -0.05        |
| 15             | 0.47         |
| 20             | 1.00         |

---

✅ **Conclusion:**
After applying **Min–Max scaling to the range [-1, 1]**, the dataset becomes:

[-1, -0.58, -0.05, 0.47, 1]



### Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform Feature Extraction using PCA. How many principal components would you choose to retain, and why?

Ans:

### **Feature Extraction using PCA:**

Given dataset features:

**[Height, Weight, Age, Gender, Blood Pressure]**

PCA can be used to **extract new features (principal components)** that capture most of the important information in the dataset while reducing dimensionality.

---

## **Step 1: Preprocess the Data:**

Before applying PCA:

1. **Convert categorical data**

   * Gender → Male = 0, Female = 1 (or One-Hot Encoding)

2. **Standardize the data**

   * Ensure all features have **mean = 0 and variance = 1**.

---

## **Step 2: Apply PCA:**

PCA transforms the 5 original features into **principal components**:

| Original Features | Principal Components |
| ----------------- | -------------------- |
| Height            | PC1                  |
| Weight            | PC2                  |
| Age               | PC3                  |
| Gender            | PC4                  |
| Blood Pressure    | PC5                  |

Each principal component is a **combination of the original variables**.

---

## **Step 3: Decide Number of Components:**

The number of components is chosen based on **Explained Variance Ratio**.

Example:

| Component | Variance Explained |
| --------- | ------------------ |
| PC1       | 45%                |
| PC2       | 25%                |
| PC3       | 15%                |
| PC4       | 10%                |
| PC5       | 5%                 |

Total variance captured:

* PC1 + PC2 + PC3 = **85%**

---

## **Step 4: Select Components:**

We would retain **3 principal components (PC1, PC2, PC3)** because they capture **around 80–90% of the total variance**, which preserves most of the important information while reducing dimensionality.

---

## **Example Transformation:**

Original dataset:

| Height | Weight | Age | Gender | BP |
| ------ | ------ | --- | ------ | -- |

After PCA:

| PC1 | PC2 | PC3 |
| --- | --- | --- |

These components represent the **major patterns in the data**.

---

## **Conclusion:**

Using PCA for feature extraction:

* The original **5 features are reduced to 3 principal components**.
* These components retain **most of the dataset’s variance (≈85%)**.
* This reduces dimensionality, improves model efficiency, and removes correlated features.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Sample dataset
data = {
    'height': [160, 170, 165, 180, 175],
    'weight': [60, 70, 65, 80, 75],
    'age': [25, 30, 28, 35, 32],
    'gender': [0, 1, 1, 0, 1],  # 0 = Male, 1 = Female
    'blood_pressure': [120, 130, 125, 140, 135]
}

df = pd.DataFrame(data)

# Step 1: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Apply PCA
pca = PCA(n_components=3)   # retain 3 principal components
principal_components = pca.fit_transform(scaled_data)

# Step 3: Create DataFrame for PCA result
pca_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2', 'PC3'])

print("Original Dataset:\n", df)
print("\nPCA Transformed Dataset:\n", pca_df)

# Explained variance
print("\nExplained Variance Ratio:", pca.explained_variance_ratio_)
```
