### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:

**Five common plots in the Seaborn library** along with their uses:

---

### 1) Line Plot (`sns.lineplot()`)

**Use:**

* Shows trends over time (time series data).
* Displays relationship between two continuous variables.
* Useful for tracking changes (e.g., sales growth per month).

**Example Use Case:** Visualizing stock prices over time.

---

### 2) Bar Plot (`sns.barplot()`)

**Use:**

* Compares categorical data.
* Shows the average (mean) of a numerical variable for different categories.
* Helpful in comparing groups.

**Example Use Case:** Comparing average marks of students in different subjects.

---

### 3) Scatter Plot (`sns.scatterplot()`)

**Use:**

* Shows relationship between two continuous variables.
* Identifies patterns, trends, or correlations.
* Detects outliers.

**Example Use Case:** Studying relationship between height and weight.

---

### 4) Histogram (`sns.histplot()`)

**Use:**

* Shows distribution of a single numerical variable.
* Displays frequency of data values.
* Helps understand spread and skewness.

**Example Use Case:** Distribution of exam scores in a class.

---

### 5) Box Plot (`sns.boxplot()`)

**Use:**

* Shows distribution using quartiles.
* Identifies median, spread, and outliers.
* Compares distributions across categories.

**Example Use Case:** Comparing salary distribution across departments.

---



### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Step 3: Create the line plot
plt.figure(figsize=(10,6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Step 4: Add title and labels
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Step 5: Show plot
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the titanic dataset
titanic = sns.load_dataset("titanic")

# -----------------------------
# Box Plot 1: pclass vs age
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='pclass', y='age')

plt.title("Passenger Class vs Age")
plt.xlabel("Passenger Class (pclass)")
plt.ylabel("Age")
plt.show()

# -----------------------------
# Box Plot 2: pclass vs fare
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='pclass', y='fare')

plt.title("Passenger Class vs Fare")
plt.xlabel("Passenger Class (pclass)")
plt.ylabel("Fare")
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Step 3: Plot histogram
plt.figure(figsize=(10,6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Step 4: Add title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Step 5: Show plot
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the iris dataset
iris = sns.load_dataset("iris")

# Step 3: Create pair plot
sns.pairplot(data=iris, hue="species")

# Step 4: Show plot
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the flights dataset
flights = sns.load_dataset("flights")

# Step 3: Pivot the dataset
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Step 4: Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")

# Step 5: Add title
plt.title("Heatmap of Flight Passengers Over Years")
plt.xlabel("Year")
plt.ylabel("Month")

# Step 6: Show plot
plt.show()
```
