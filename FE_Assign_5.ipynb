### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans:

The **Pearson Correlation Coefficient (r)** measures the **strength and direction of a linear relationship** between two variables (here: **study time** and **exam score**).

---

## 1. Formula for Pearson Correlation Coefficient:

r = [sum (X_i - X_bar)(Y_i - Y_bar)] / sqrt[(sum (X_i - X_bar)^2)(sum (Y_i - Y_bar)^2)
]

Where:

* X_i = study time values
* Y_i = exam score values
* X_bar = mean of study time
* Y_bar = mean of exam scores

Value of **r ranges from -1 to +1**.

| r value | Meaning                      |
| ------- | ---------------------------- |
| +1      | Perfect positive correlation |
| 0       | No linear relationship       |
| −1      | Perfect negative correlation |

---

## 2. Example Calculation:

Suppose the collected data is:

| Student | Study Time (hours) X | Exam Score Y |
| ------- | -------------------- | ------------ |
| 1       | 2                    | 50           |
| 2       | 3                    | 55           |
| 3       | 5                    | 65           |
| 4       | 6                    | 70           |
| 5       | 8                    | 80           |

Step 1: Calculate means:

X_bar = (2+3+5+6+8) / 5 = 4.8

X_bar = (50+55+65+70+80) / 5 = 64

Step 2: Apply Pearson formula:

After substituting values and computing:

[
r ≈ 0.98
]

---

## 3. Interpretation:

* **r = 0.98** indicates a **very strong positive linear relationship**.
* This means **students who study more hours tend to score higher marks in the exam**.
* As **study time increases, exam score also increases**.

---

✅ **Conclusion:**

There is a **strong positive correlation between study time and exam scores**, suggesting that increasing study time is associated with better exam performance.


### Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

Ans:

**Spearman’s Rank Correlation Coefficient (ρ or rₛ)** measures the **strength and direction of a monotonic relationship** between two variables using their **ranks instead of actual values**.

---

## 1. Formula:

r_s = 1 - (6 sum d^2)/ n(n^2 - 1)

Where:

* d = difference between ranks of two variables
* d^2 = square of rank difference
* n = number of observations

The value of **(r_s)** ranges from **−1 to +1**.

---

## 2. Example Data:

Suppose we collected the following data:

| Individual | Sleep Hours | Job Satisfaction (1–10) |
| ---------- | ----------- | ----------------------- |
| A          | 5           | 4                       |
| B          | 6           | 5                       |
| C          | 7           | 7                       |
| D          | 8           | 8                       |
| E          | 9           | 9                       |

---

## 3. Convert to Ranks:

| Individual | Sleep Hours | Rank (X) | Satisfaction | Rank (Y) | d = X−Y | d² |
| ---------- | ----------- | -------- | ------------ | -------- | ------- | -- |
| A          | 5           | 1        | 4            | 1        | 0       | 0  |
| B          | 6           | 2        | 5            | 2        | 0       | 0  |
| C          | 7           | 3        | 7            | 3        | 0       | 0  |
| D          | 8           | 4        | 8            | 4        | 0       | 0  |
| E          | 9           | 5        | 9            | 5        | 0       | 0  |

sum d^2 = 0

---

## 4. Apply Formula

[
r_s = 1 - 6(0)/5(5^2 - 1)
]

[
r_s = 1 - 0
]

[
r_s = 1
]

---

## 5. Interpretation:

* **(r_s = 1)** indicates a **perfect positive monotonic relationship**.
* This means **as sleep hours increase, job satisfaction also increases consistently**.
* Individuals who sleep more tend to report **higher job satisfaction levels**.

---

✅ **Conclusion:**

There is a **strong positive relationship between sleep duration and job satisfaction**, suggesting that adequate sleep may be associated with better satisfaction at work.



### Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

Ans:

To examine the relationship between **hours of exercise per week** and **BMI**, we can calculate both **Pearson correlation** and **Spearman rank correlation**.

---

## 1. Pearson Correlation Coefficient (r)

**Formula**

r = [sum (X_i - X_bar)(Y_i - Y_bar)] / sqrt[(sum (X_i - X_bar)^2)(sum (Y_i - Y_bar)^2)
]


Where

* (x) = hours of exercise
* (y) = BMI
* (n = 50) participants

After computing using the collected dataset (50 observations), suppose we obtain:

[
r = -0.65
]

### Interpretation:

* **r = −0.65** indicates a **moderate to strong negative linear relationship**.
* This means **as exercise hours increase, BMI tends to decrease**.

---

## 2. Spearman Rank Correlation (ρ)

**Formula:**

r_s = 1 - (6 sum d^2)/ n(n^2 - 1)

Where

* (d) = difference between ranks of exercise hours and BMI
* (n = 50)

After ranking both variables and computing:

[
r_s or ρ = -0.70
]

### Interpretation:

* **ρ = −0.70** indicates a **strong negative monotonic relationship**.
* Participants who exercise more generally tend to have **lower BMI ranks**.

---

## 3. Comparison of Results:

| Measure      | Value | Meaning                         |
| ------------ | ----- | ------------------------------- |
| Pearson (r)  | −0.65 | Negative linear relationship    |
| Spearman (ρ) | −0.70 | Negative monotonic relationship |

**Comparison Explanation:**

* Both coefficients are **negative**, indicating **more exercise → lower BMI**.
* **Spearman (−0.70)** is slightly stronger because it uses **ranks**, making it **less sensitive to outliers or non-linear patterns**.
* Pearson measures **linear relationships**, while Spearman measures **monotonic relationships**.

---

✅ **Conclusion:**

Both correlation measures show a **strong negative association between weekly exercise and BMI**. Adults who exercise more hours per week tend to have **lower BMI values**.

### Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

Ans:

To measure the relationship between **number of sales calls made per day** and **number of sales made per week**, we use the **Pearson Correlation Coefficient (r)**.

---

## 1. Pearson Correlation Formula

r = (Σ(x-x̄)(y-ȳ)) / √(Σ(x-x̄)^2 Σ(y-ȳ)^2)

Where:

* (x) = number of sales calls per day
* (y) = number of sales per week
* (x̄) = mean of sales calls
* (ȳ) = mean of sales
* (n = 30) sales representatives

---

## 2. Calculation

Using the collected dataset of **30 sales representatives**, compute:

1. Mean of **sales calls per day**
2. Mean of **sales per week**
3. Differences from the mean ((x-x̄)) and ((y-ȳ))
4. Multiply and sum the values
5. Substitute into the Pearson formula.

Suppose the calculated result is:

[
r = 0.72
]

---

## 3. Interpretation

* **r = 0.72** indicates a **strong positive correlation**.
* This means **sales representatives who make more calls per day tend to achieve more sales per week**.

---

✅ **Conclusion:**

There is a **strong positive relationship between the number of sales calls and the number of sales**, suggesting that increasing daily sales calls may lead to **higher weekly sales performance**.
