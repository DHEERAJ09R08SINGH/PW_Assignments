Ans:


## 🎯 Goal:

> Predict the **actual house price as accurately as possible**

---

# 🔷 1. Why MSE is Better

### 📌 Formula:

MSE=1/n​∑(yi​−y^​i​)^2

### ✅ Key Reasons:

* Directly measures **prediction error**
* Penalizes **large mistakes more heavily**
* Focuses on **numerical accuracy (actual price difference)**

👉 Example:

* Predict ₹50L instead of ₹1Cr → large error → heavily penalized

---

# 🔷 2. Why NOT R² (Alone)

### 📌 What R² does:

* Measures **how well model explains variance**
* NOT actual prediction error

👉 Problem:

* Two models can have same R² but different **price errors**

### 📊 Example:

| Model | R²   | Error    |
| ----- | ---- | -------- |
| A     | 0.90 | ₹2 lakh  |
| B     | 0.90 | ₹10 lakh |

👉 Both look same in R², but A is clearly better

---

# 🔷 3. Key Difference

| Metric | Focus                       |
| ------ | --------------------------- |
| MSE    | Actual error (₹ difference) |
| R²     | Explained variance (%)      |

---

# 🔷 4. Final Recommendation

✔️ Use **MSE (or RMSE)** when:

* You care about **accurate price prediction**
* Errors must be minimized in real units

✔️ Use **R²** when:

* Comparing models
* Understanding overall fit

---



👉 *MSE is more appropriate because it directly measures the difference between predicted and actual house prices and penalizes large errors. Since the goal is to achieve accurate price prediction, minimizing prediction error is more important than explaining variance, which is what R² measures.*




### Q3. You have a dataset with a significant number of outliers and are trying to select an appropriate regression metric to use with your SVM model. Which metric would be the most appropriate in this scenario?

Ans:

✔️ **MAE (Mean Absolute Error)** is the most appropriate metric.

---

# 🔷 Why MAE is Best for Outliers

### 📌 Formula:

MAE=1/n​∑∣yi​−y^​i​∣

### ✅ Key Reason:

* MAE uses **absolute error**, not squared error
* So it **does NOT heavily penalize outliers**

---

# 🔷 Comparison with Other Metrics

| Metric         | Effect of Outliers  |
| -------------- | ------------------- |
| **MAE**        | Less sensitive ✅    |
| **MSE / RMSE** | Highly sensitive ❌  |
| **R²**         | Can be misleading ❌ |

---

# 🔷 Example

Suppose:

* Actual price = ₹50L
* Predicted = ₹5Cr (outlier case)

👉 Error = ₹4.5Cr

* **MSE** → squares it → huge impact
* **MAE** → keeps it linear → controlled impact

---

# 🔷 Intuition

* **MSE/RMSE** → punish big errors heavily
* **MAE** → treat all errors equally

👉 So MAE is **robust to outliers**

---

# 🔷 When to Use MAE

✔️ Dataset has:

* Noisy values
* Extreme prices (luxury houses, rare cases)
* Data entry errors

---



👉 *MAE is the most appropriate metric when the dataset contains a significant number of outliers because it measures absolute error and is less sensitive to extreme values. Unlike MSE or RMSE, it does not excessively penalize large errors, making it more robust in such scenarios.*




### Q4. You have built an SVM regression model using a polynomial kernel and are trying to select the best metric to evaluate its performance. You have calculated both MSE and RMSE and found that both values are very close. Which metric should you choose to use in this case?

Ans:


✔️ **Choosing RMSE (Root Mean Squared Error)**

---

# 🔷 Why RMSE is Better (when MSE ≈ RMSE comparison)

Even if both are “close” in value, **RMSE is more interpretable**.

---

# 🔷 1. Key Difference

| Metric | Unit                 |
| ------ | -------------------- |
| MSE    | squared units (₹²) ❌ |
| RMSE   | original units (₹) ✅ |

👉 RMSE converts error back to **same unit as house price**

---

# 🔷 2. Why This Matters

### 📌 Example:

* MSE = 25,00,00,000 → hard to interpret
* RMSE = 50,000 → clear meaning

👉 RMSE tells:

> “Average prediction error is ₹50,000”

---

# 🔷 3. When Values Are Close

* RMSE = √MSE
* So they will always be mathematically related

👉 But:

* **MSE → good for optimization**
* **RMSE → good for interpretation & reporting**

---

# 🔷 4. Final Recommendation

✔️ Use **RMSE** because:

* Easy to understand
* Same unit as target variable
* Better for real-world decisions

---


👉 *When both MSE and RMSE values are close, RMSE should be preferred because it is expressed in the same units as the target variable, making it easier to interpret. It provides a more intuitive understanding of the prediction error compared to MSE, which is in squared units.*



### Q5. You are comparing the performance of different SVM regression models using different kernels (linear, polynomial, and RBF) and are trying to select the best evaluation metric. Which metric would be most appropriate if your goal is to measure how well the model explains the variance in the target variable?

Ans:


✔️ **R² Score (Coefficient of Determination)** is the most appropriate metric.

---

# 🔷 Why R² is Best Here

### 🎯 Your Goal:

> Measure how well the model **explains variance** in house prices

👉 R² is **specifically designed** for this purpose.

---

# 🔷 1. What R² Measures

### 📌 Formula:

R^2=1−(​SS_res​​/SS_tot)

### ✅ Meaning:

* Shows how much of the **variation in target variable** is explained by the model

---

# 🔷 2. Interpretation

| R² Value | Meaning               |
| -------- | --------------------- |
| 1.0      | Perfect prediction    |
| 0.0      | No explanatory power  |
| < 0      | Worse than mean model |

👉 Example:

* R² = 0.85 → model explains **85% variance in house prices**

---

# 🔷 3. Why Not MSE / RMSE?

| Metric     | Limitation                              |
| ---------- | --------------------------------------- |
| MSE / RMSE | Measure error, not variance explanation |

👉 They tell:

* “How wrong is the model?”
  👉 But NOT:
* “How well does it explain data variation?”

---

# 🔷 4. Best Use Case

✔️ Comparing models with:

* Different kernels (Linear vs Poly vs RBF)
* Different complexities

👉 R² helps identify **which model captures patterns better**

---



👉 *R² score is the most appropriate metric because it measures how well the model explains the variance in the target variable. It is particularly useful for comparing different SVM regression models with different kernels, as it indicates the proportion of variance captured by each model.*
